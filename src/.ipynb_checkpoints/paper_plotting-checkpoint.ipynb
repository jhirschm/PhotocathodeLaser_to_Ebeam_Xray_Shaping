```python
import os
import re
import glob
import copy
from datetime import datetime
from typing import Union, List, Tuple

import numpy as np
import matplotlib.pyplot as plt
import matplotlib

import h5py
import cv2

from scipy.ndimage import gaussian_filter1d, shift
from scipy.signal import correlate, find_peaks
from scipy import signal

import random


matplotlib.rcParams['pdf.fonttype'] = 42
plt.rcParams["font.size"] = 15
plt.rcParams["xtick.color"] = 'black'
plt.rcParams["ytick.color"] = 'black'
plt.rcParams["axes.edgecolor"] = 'black'
plt.rcParams["axes.linewidth"] = 1
```

```python
correlation_dtype = np.dtype([
    ("type", "U20"),
    ("instrument", "U50"),
    ("date", "U10"),
    ("timestamp", int),       # HHMMSS parsed from filename
    ("filename", "U100"),
    ("charge", object),       # (nshots,) array filled later
    ("valid_charge", bool),
    ("image", object),        # (nshots, H, W) array filled later
    ("valid_image", bool)
])

def initialize_correlation_struct(folder: str) -> np.ndarray:
    """Index `CorrelationPlot-*` files into a structured array."""
    entries = []
    for fname in os.listdir(folder):
        if not fname.endswith(".mat") or not fname.startswith("CorrelationPlot-"):
            continue
        match = re.match(
            r"(?P<type>\w+)-(?P<instr>[\w_]+)-(?P<date>\d{4}-\d{2}-\d{2})-(?P<hh>\d{2})(?P<mm>\d{2})(?P<ss>\d{2})\.mat",
            fname
        )
        if match:
            hhmmss = int(match["hh"] + match["mm"] + match["ss"])
            entries.append((
                match["type"], match["instr"], match["date"], hhmmss, fname,
                None, False, None, False
            ))
    return np.array(entries, dtype=correlation_dtype)

def initialize_trex_struct(folder: str) -> np.ndarray:
    """Index `TREXB-*` files (calibration metadata) into a structured array."""
    entries = []
    for fname in os.listdir(folder):
        if not fname.endswith(".mat") or not fname.startswith("TREXB-"):
            continue
        match = re.match(
            r"(?P<type>\w+)-(?P<instr>[\w_]+)-(?P<date>\d{4}-\d{2}-\d{2})-(?P<hh>\d{2})(?P<mm>\d{2})(?P<ss>\d{2})\.mat",
            fname
        )
        if match:
            hhmmss = int(match["hh"] + match["mm"] + match["ss"])
            entries.append((
                match["type"], match["instr"], match["date"], hhmmss, fname,
                None, False, None, False
            ))
    return np.array(entries, dtype=correlation_dtype)

def initialize_diag0_525_struct(folder: str) -> np.ndarray:
    """Index `diag0_525` files (screen resolution helper) into a structured array."""
    entries = []
    pattern = re.compile(
        r"(?P<type>[\w\-]+)-(?P<date>\d{4}-\d{2}-\d{2})-(?P<hh>\d{2})(?P<mm>\d{2})(?P<ss>\d{2})\.mat",
        re.IGNORECASE
    )
    for fname in os.listdir(folder):
        if ".mat" not in fname.lower() or "diag0_525" not in fname.lower():
            continue
        match = pattern.search(fname)
        if match:
            hhmmss = int(match["hh"] + match["mm"] + match["ss"])
            entries.append((
                match["type"], None, match["date"], hhmmss, fname,
                None, False, None, False
            ))
    return np.array(entries, dtype=correlation_dtype)
```

```python
def deref_utf16_string(f, ref):
    return f[ref][()].tobytes()[::2].decode("utf-8").strip()

def deref_scalar(f, ref):
    val = f[ref][()]
    return val.item() if hasattr(val, "item") else val

def populate_correlation_entries(struct: np.ndarray,
                                 folder: str,
                                 timestamps: Union[int, List[int], Tuple[int, int]],
                                 target_pv: str = "BPMS:GUNB:314:FW:CHRG"):
    """Populate image stacks + target PV readback for the requested timestamp(s)."""
    if isinstance(timestamps, int):
        timestamps = [timestamps]
    elif isinstance(timestamps, tuple) and len(timestamps) == 2:
        start, end = timestamps
        timestamps = [t for t in struct["timestamp"] if start <= t <= end]

    for idx in np.where(np.isin(struct["timestamp"], timestamps))[0]:
        row = struct[idx]
        full_path = os.path.join(folder, row["filename"])

        try:
            with h5py.File(full_path, "r") as f:
                # --- images ---
                img_refs = f["data"]["dataList"]["img"]
                npoints = img_refs.shape[0]
                image_stack = np.stack([f[img_refs[j, 0]][()] for j in range(npoints)])
                struct["image"][idx] = image_stack
                struct["valid_image"][idx] = True

                # --- PV readbacks ---
                val_refs = f["data/readPV/val"]
                name_refs = f["data/readPV/name"][0, 0, :]
                _, _, nreadPV = val_refs.shape

                target_index = None
                for i in range(nreadPV):
                    name = deref_utf16_string(f, name_refs[i])
                    if name.strip() == target_pv:
                        target_index = i
                        break
                if target_index is None:
                    print(f"[Skip] PV '{target_pv}' not found in {row['filename']}")
                    continue

                charges = np.array([deref_scalar(f, val_refs[j, 0, target_index]) for j in range(npoints)])
                struct["charge"][idx] = charges
                struct["valid_charge"][idx] = True

        except Exception as e:
            print(f"[Error] {row['filename']}: {e}")

def get_charge_and_image_by_timestamp(struct: np.ndarray, timestamp: int):
    idx_arr = np.where(struct["timestamp"] == timestamp)[0]
    if len(idx_arr) == 0:
        raise ValueError(f"Timestamp {timestamp} not found.")
    idx = idx_arr[0]
    charge = struct["charge"][idx]
    images = struct["image"][idx]
    valid_charge = struct["valid_charge"][idx]
    valid_image = struct["valid_image"][idx]
    if charge is None or images is None:
        raise ValueError(f"Entry at timestamp {timestamp} has not been populated yet.")
    return charge, images, valid_charge, valid_image
```

```python
def process_tcav_image_v1(
    image: np.ndarray,
    threshold: float = 10.0,
    kernel_size: int = 3,
    erode_iters: int = 1,
    dilate_iters: int = 1,
) -> np.ndarray:

    """
    TCAV preprocessing:
      1. Erosion (remove speckle)
      2. Threshold-to-zero (background removal)
      3. Dilation (restore blob size)

    Designed so that:
      - threshold controls background removal
      - morphology only smooths, not reshapes
    """

    img = image.astype(np.float32)

    kernel = np.ones((kernel_size, kernel_size), np.float32)

    if erode_iters > 0:
        img = cv2.erode(img, kernel, iterations=erode_iters)

    # Threshold-to-zero: keeps intensities above threshold
    _, img = cv2.threshold(img, threshold, 0, cv2.THRESH_TOZERO)

    if dilate_iters > 0:
        img = cv2.dilate(img, kernel, iterations=dilate_iters)

    return img
```

```python
def get_tcavAxes_conversion_ticks(
    energy,                 # MeV
    streak,                 # mm/deg
    dispersion,             # (see note in markdown cell)
    cavity_freq=2.856e9,    # Hz
    resolution=27.47,       # μm/pixel
    c=3e8
):
    print("Expects: streak in mm/deg, energy in MeV, resolution in μm/px, dispersion in m")
    # seconds per degree of RF phase
    lambda_band = c / cavity_freq
    tdeg = lambda_band / 360 / c   # = 1/(360 f)

    res_m_per_px = resolution * 1e-6
    streak_m_per_deg = streak * 1e-3

    px2fs = (tdeg / streak_m_per_deg) * res_m_per_px * 1e15

    # as in your code (may be normalized dispersion; revisit later if needed)
    px2mev = (res_m_per_px) / dispersion * energy

    xticks = np.array([100, 300, 500, 700])
    yticks = np.array([60, 180, 300, 420])

    print("Cavity, dispersion, energy, resolution, streak")
    print(cavity_freq)
    print(dispersion)
    print(energy)
    print(resolution)
    print(streak)
    return px2fs, px2mev, xticks, yticks
```

```python
def moving_average(x: np.ndarray, w: int) -> np.ndarray:
    return np.convolve(x, np.ones(w), 'valid') / w
```

```python
def detect_edge_clipping_from_uncentered_projection(
    img2d_uncentered,
    edge_bins_frac=0.10,
    edge_bins_min=10,
    edge_bins_max=None,
    # --- NEW: primary decision uses edge-mass fraction ---
    clip_fraction_threshold=0.18,      # start ~0.15-0.25
    # --- Optional asymmetry guard (rarely needed) ---
    require_asymmetry=False,
    asymmetry_ratio_threshold=1.5,     # only used if require_asymmetry=True
    # --- Baseline removal on projection (important) ---
    proj_baseline_percentile=5.0,      # subtract p5 of proj to kill background pedestal
    # stats for debug only
    robust_stat="mean",                # "mean" or "median" for edge-bin stats
    debug=False,
    debug_id=None,
):
    """
    Detect x-edge clipping on an UNCENTERED 2D image by looking at the 1D x-projection.

    Key idea:
      - Work AFTER your erosion/dilation/threshold pipeline (recommended).
      - Build x-projection: proj = sum(img2d, axis=0)
      - Remove a pedestal on the 1D projection: proj = max(proj - p{proj_baseline_percentile}, 0)
      - Compute edge mass fractions:
            left_frac  = sum(proj[:edge_bins]) / sum(proj)
            right_frac = sum(proj[-edge_bins:]) / sum(proj)
      - Decide clipped if max(left_frac, right_frac) > clip_fraction_threshold.

    Returns:
      clipped (bool), info (dict)
    """
    A = np.nan_to_num(np.asarray(img2d_uncentered, float), nan=0.0, posinf=0.0, neginf=0.0)

    # If erosion/dilation returns nonnegative already, great.
    # If not, clip negatives (important after bk subtraction).
    A = np.maximum(A, 0.0)

    proj = A.sum(axis=0)  # x projection
    w = proj.size
    if w == 0:
        return False, {"reason": "empty_projection"}

    total_raw = float(np.sum(proj))
    if total_raw <= 0:
        return False, {"reason": "zero_signal"}

    # --- baseline subtract on 1D projection (kills broad pedestal that ruins ratios) ---
    base = float(np.percentile(proj, proj_baseline_percentile)) if proj_baseline_percentile is not None else 0.0
    proj_bs = proj - base
    proj_bs[proj_bs < 0] = 0.0

    total = float(np.sum(proj_bs))
    if total <= 0:
        return False, {"reason": "zero_signal_after_baseline", "baseline": base}

    # choose number of edge bins
    edge_bins = int(round(edge_bins_frac * w))
    edge_bins = max(edge_bins_min, edge_bins)
    if edge_bins_max is not None:
        edge_bins = min(edge_bins_max, edge_bins)
    edge_bins = min(edge_bins, w // 2)  # safety

    left_bins = np.arange(0, edge_bins)
    right_bins = np.arange(w - edge_bins, w)

    left_mass = float(np.sum(proj_bs[left_bins]))
    right_mass = float(np.sum(proj_bs[right_bins]))

    left_frac = left_mass / total
    right_frac = right_mass / total

    # --- optional: keep the old-style edge stat ratios for debug/guard only ---
    stat_fn = np.mean if robust_stat == "mean" else np.median
    left_stat = float(stat_fn(proj_bs[left_bins]))
    right_stat = float(stat_fn(proj_bs[right_bins]))
    eps = 1e-12
    lr_ratio = (left_stat + eps) / (right_stat + eps)
    rl_ratio = (right_stat + eps) / (left_stat + eps)

    # --- decision ---
    clipped_side = None
    clipped = False

    edge_frac = max(left_frac, right_frac)
    if edge_frac > clip_fraction_threshold:
        clipped = True
        clipped_side = "left" if left_frac > right_frac else "right"

        if require_asymmetry:
            # only keep clip if edges are also meaningfully asymmetric
            if not (lr_ratio >= asymmetry_ratio_threshold or rl_ratio >= asymmetry_ratio_threshold):
                clipped = False
                clipped_side = None

    info = dict(
        w=w,
        edge_bins=edge_bins,
        edge_bins_frac=edge_bins_frac,
        clip_fraction_threshold=clip_fraction_threshold,
        proj_baseline_percentile=proj_baseline_percentile,
        baseline=base,
        total_mass=float(total),
        left_mass=left_mass,
        right_mass=right_mass,
        left_frac=float(left_frac),
        right_frac=float(right_frac),
        edge_frac=float(edge_frac),
        robust_stat=robust_stat,
        left_stat=left_stat,
        right_stat=right_stat,
        lr_ratio=float(lr_ratio),
        rl_ratio=float(rl_ratio),
        require_asymmetry=require_asymmetry,
        asymmetry_ratio_threshold=asymmetry_ratio_threshold,
        clipped=clipped,
        clipped_side=clipped_side,
        # for external plotting/debug if you want:
        proj_bs=proj_bs,
        left_bins=left_bins,
        right_bins=right_bins,
    )

    if debug:
        tag = f"[{debug_id}] " if debug_id is not None else ""
        print(f"{tag}edge_bins={edge_bins} w={w} total_mass={total:.3g} baseline(p{proj_baseline_percentile})={base:.3g}")
        print(f"{tag}left_frac={left_frac:.5f} right_frac={right_frac:.5f} edge_frac={edge_frac:.5f} thr={clip_fraction_threshold}")
        print(f"{tag}left_stat={left_stat:.4f} right_stat={right_stat:.4f} ratios L/R={lr_ratio:.3f} R/L={rl_ratio:.3f}")
        print(f"{tag}DECISION: clipped={clipped} side={clipped_side}")

    return clipped, info


def prep_xtcav(
    target_timestamp,
    trex_stamp,
    hard_code_streak=7.37,
    image_summation_threshold=1000,
    moving_average_window=30,
    y_moving_average_window=None,
    threshold=5,
    every_other_plot_number=1,
    bk_subtract=False,
    bk_timestamp=None,

    # --- auto-exclusion on UNCENTERED projection ---
    recommend_excludes=True,
    detect_debug=False,
    detect_debug_max_print=10,
    detect_edge_bins_frac=0.10,
    detect_edge_bins_min=10,
    detect_clip_fraction_threshold=0.18,
    detect_proj_baseline_percentile=5.0,
    detect_require_asymmetry=False,
    detect_asymmetry_ratio_threshold=1.5,
    detect_robust_stat="mean",

    # --- plotting ---
    plot_uncentered=True,
    color="jet",
    x_lim=(-400, 400),
    y_lim=(10, 50),

    resolution_hardcoded = None
):
    """
    Returns:
      proj_list, charge_list, idx_list, x_phys,
      recommended_exclude_raw, recommended_meta, index_map

    Notes:
      - idx_list[k] is RAW index into images for proj_list[k].
      - detection happens AFTER process_tcav_image_v1 (erosion/dilation/etc),
        using baseline-subtracted 1D projection -> stable across threshold choices.
    """
    filtered_entries = trex_struct_sorted[trex_struct_sorted["timestamp"] == trex_stamp]
    filename = filtered_entries["filename"][0]
    trex = beam_time_dir + filename

    if y_moving_average_window is None:
        y_moving_average_window = moving_average_window

    # --- TREX meta ---
    with h5py.File(trex, "r") as f:
        dispersion = np.array(f["data/dispersion"]).item()
        streak = np.array(f["data/streak"]).item()
        init_streak = np.array(f["data/init_streak"]).item()
        charge0 = np.array(f["data/charge"]).item()
        mean_energy = np.array(f["data/mean_erg"]).item()
        px2um = np.array(f["data/px2um"]).item()
        sync = np.array(f["data/sync"]).item()
    print(dispersion, streak, init_streak, charge0, mean_energy, px2um, sync)
    if resolution_hardcoded is not None:
        px2um = resolution_hardcoded
    xtcav_params = dict(
        resolution=px2um,
        dispersion=dispersion * 1e-6 * -1,
        streak=hard_code_streak,
        energy=mean_energy,
        cavity_frequency=11.424e9,
    )

    # --- load shots ---
    populate_correlation_entries(corr_struct_sorted, beam_time_dir, target_timestamp)
    charges, images, valid_charge, valid_image = get_charge_and_image_by_timestamp(
        corr_struct_sorted, target_timestamp
    )

    # --- optional background subtraction ---
    if bk_subtract and (bk_timestamp is not None):
        populate_correlation_entries(corr_struct_sorted, beam_time_dir, bk_timestamp)
        _, bk_images, _, _ = get_charge_and_image_by_timestamp(corr_struct_sorted, bk_timestamp)
        images = images - np.mean(bk_images, axis=0)

    # --- calibration ---
    px2fs, px2mev, xticks, yticks = get_tcavAxes_conversion_ticks(
        xtcav_params["energy"],
        xtcav_params["streak"],
        xtcav_params["dispersion"],
        xtcav_params["cavity_frequency"],
        xtcav_params["resolution"],
        3e8,
    )

    # --- sorting by charge defines projection index space ---
    sorted_indices = np.argsort(charges)

    proj_list = []
    charge_list = []
    idx_list = []

    recommended_exclude_raw = []
    recommended_meta = {}

    # debug counters
    n_print_flag = 0
    n_print_ok = 0

    m = 0
    x_phys = None

    for i in sorted_indices:
        # 1) your normal morphology/threshold pipeline
        filtered = process_tcav_image_v1(images[i, :, :], threshold=threshold)

        # uncentered representation used for detection + optional plotting
        img_uncentered = filtered.T  # shape (h, w)

        # 2) image-summation threshold gate (after processing)
        img_sum = float(np.sum(img_uncentered))
        if img_sum < image_summation_threshold:
            print("Image does not contain a signal; Index " + str(i))
            continue

        # 3) detect clipping AFTER erosion/dilation (stable)
        clipped = False
        info = None
        if recommend_excludes:
            debug_this = False
            if detect_debug:
                debug_this = (n_print_flag < detect_debug_max_print) or (n_print_ok < detect_debug_max_print)

            clipped, info = detect_edge_clipping_from_uncentered_projection(
                img2d_uncentered=img_uncentered,
                edge_bins_frac=detect_edge_bins_frac,
                edge_bins_min=detect_edge_bins_min,
                clip_fraction_threshold=detect_clip_fraction_threshold,
                proj_baseline_percentile=detect_proj_baseline_percentile,
                require_asymmetry=detect_require_asymmetry,
                asymmetry_ratio_threshold=detect_asymmetry_ratio_threshold,
                robust_stat=detect_robust_stat,
                debug=debug_this,
                debug_id=f"raw_idx={i}",
            )

            if debug_this:
                if clipped:
                    n_print_flag += 1
                else:
                    n_print_ok += 1

            if clipped:
                recommended_exclude_raw.append(int(i))
                recommended_meta[int(i)] = info

        # 4) scale by charge (your original behavior)
        charge_i = float(charges[i])
        img_normalized_uncentered = (img_uncentered / img_sum) 
        scaler_xproj = charge_i * 1e-12 / (np.sum(img_normalized_uncentered) * px2fs * 1e-15) # Charge in pc over time step in fs
        scaler_yproj = charge_i * 1e-12 / (np.sum(img_normalized_uncentered) * px2mev * 1e6)

        h, w = img_normalized_uncentered.shape

        # centroid for recentering
        x_idx = np.arange(w)
        y_idx = np.arange(h)
        total = img_normalized_uncentered.sum()
        x_c = (img_normalized_uncentered.sum(axis=0) @ x_idx) / total
        y_c = (img_normalized_uncentered.sum(axis=1) @ y_idx) / total
        dx = int(round(w / 2 - x_c))
        dy = int(round(h / 2 - y_c))

        # recenter for projections
        img_centered = np.roll(np.roll(img_normalized_uncentered, shift=dy, axis=0), shift=dx, axis=1)

        # physical axes (centered convention)
        x_phys = (np.arange(w) - w / 2) * px2fs
        y_phys = (np.arange(h) - h / 2) * px2mev

        # x projection from centered image
        x_projection = np.sum(img_centered*scaler_xproj, axis=0)
        x_projection_smooth = moving_average(x_projection, moving_average_window)
        pad = (x_projection.shape[0] - x_projection_smooth.shape[0]) // 2
        x_projection_smooth_padded = np.pad(
            x_projection_smooth,
            (pad, x_projection.shape[0] - x_projection_smooth.shape[0] - pad),
            mode="edge",
        )

        # y projection (optional)
        y_projection = np.sum(img_centered*scaler_yproj, axis=1)
        y_projection_smooth = moving_average(y_projection, y_moving_average_window)
        pad = (y_projection.shape[0] - y_projection_smooth.shape[0]) // 2
        y_projection_smooth_padded = np.pad(
            y_projection_smooth,
            (pad, y_projection.shape[0] - y_projection_smooth.shape[0] - pad),
            mode="edge",
        )

        # optional plotting
        if plot_uncentered and (m == 0):
            proj_idx = len(proj_list)

            fig, ax = plt.subplots(figsize=(8, 6))

            x_phys_unc = (np.arange(w) - w / 2) * px2fs
            y_phys_unc = (np.arange(h) - h / 2) * px2mev

            im = ax.imshow(
                img_normalized_uncentered,
                cmap=color,
                origin="upper",
                extent=[x_phys_unc[0], x_phys_unc[-1], y_phys_unc[-1], y_phys_unc[0]],
                aspect="auto",
                rasterized=False,
            )

            ax.set_title(
                f"raw_idx {i} | proj_idx {proj_idx} | charge={charge_i:.3f} | clipped={clipped}"
            )
            ax.set_xlim(*x_lim)
            ax.set_ylim(*y_lim)
            ax.set_xlabel("Time (fs)")
            ax.set_ylabel("Delta Energy (MeV)")

            # overlay the projection used for detection (baseline-subtracted)
            proj_bs = info["proj_bs"] if (info is not None and "proj_bs" in info) else img_uncentered.sum(axis=0)
            proj_bs_n = proj_bs / (np.max(proj_bs) + 1e-12)

            axr = ax.twinx()
            axr.plot(x_phys_unc, proj_bs_n, color="red", lw=0.8, alpha=0.9)
            axr.set_ylabel("Detection proj (norm)", color="red")
            axr.tick_params(axis="y", colors="red")
            axr.set_ylim(0, 1.05)

            cbar = fig.colorbar(im, ax=ax, pad=0.02, fraction=0.05)
            cbar.set_label("Intensity (a.u.)")
            plt.show()

        m += 1
        if m > every_other_plot_number:
            m = 0

        proj_list.append(x_projection_smooth_padded.astype(float))
        charge_list.append(charge_i)
        idx_list.append(int(i))

    recommended_exclude_raw = sorted(set(recommended_exclude_raw))

    proj_to_raw = {k: int(raw) for k, raw in enumerate(idx_list)}
    raw_to_proj = {int(raw): k for k, raw in enumerate(idx_list)}
    index_map = {"proj_to_raw": proj_to_raw, "raw_to_proj": raw_to_proj}

    return (
        proj_list,
        charge_list,
        idx_list,
        x_phys,
        recommended_exclude_raw,
        recommended_meta,
        index_map,
    )


```

```python
def estimate_xtcav_background_stats(
    target_timestamp: int,
    bk_subtract: bool = False,
    bk_timestamp: int = None,
    # ROI in ORIGINAL "images[i,:,:]" indexing
    # (you suggested x=924..1023, y=0..99 for bottom-right)
    roi_x: tuple = (924, 1024),   # x start inclusive, end exclusive
    roi_y: tuple = (0, 100),      # y start inclusive, end exclusive
    # which images to use (after sorting etc). Default: use all
    use_indices: list = None,
    max_images: int = None,
    # robust stats
    robust: bool = True,
    # debug
    return_samples: bool = False,
    verbose: bool = True,
):
    """
    Loads XTCAV images using the same timestamp-based mechanism as prep_xtcav,
    optionally background-subtracts using bk_timestamp, then computes background
    ROI statistics aggregated over MANY images.

    IMPORTANT:
      - This function works on the raw image orientation images[i,:,:]
        (before your later filtered.T transpose).
      - The ROI coordinates (roi_x, roi_y) refer to that raw orientation.
      - If you want ROI selection in the transposed convention, flip x/y or
        apply transpose before ROI slicing.

    Returns:
      stats (dict) with fields:
        - n_images_used
        - roi_x, roi_y
        - mean, median, std
        - mad, sigma_mad (robust sigma estimate)
        - p01, p05, p95, p99
      optionally 'samples' (1D array of ROI pixels) if return_samples=True
    """
    # --- load shots ---
    populate_correlation_entries(corr_struct_sorted, beam_time_dir, target_timestamp)
    charges, images, valid_charge, valid_image = get_charge_and_image_by_timestamp(
        corr_struct_sorted, target_timestamp
    )

    if images is None or len(images) == 0:
        raise ValueError(f"No images found for target_timestamp={target_timestamp}")

    images_use = images.astype(np.float32, copy=False)

    # --- optional background subtraction (match prep_xtcav pattern) ---
    if bk_subtract:
        if bk_timestamp is None:
            raise ValueError("bk_subtract=True but bk_timestamp is None")
        populate_correlation_entries(corr_struct_sorted, beam_time_dir, bk_timestamp)
        _, bk_images, _, _ = get_charge_and_image_by_timestamp(corr_struct_sorted, bk_timestamp)
        if bk_images is None or len(bk_images) == 0:
            raise ValueError(f"No background images found for bk_timestamp={bk_timestamp}")
        bk_mean = np.mean(bk_images.astype(np.float32, copy=False), axis=0)
        images_use = images_use - bk_mean

    n_total = images_use.shape[0]

    # choose which images to use
    if use_indices is None:
        idx = np.arange(n_total)
    else:
        idx = np.array([i for i in use_indices if 0 <= i < n_total], dtype=int)

    if max_images is not None and len(idx) > max_images:
        idx = idx[:max_images]

    x0, x1 = roi_x
    y0, y1 = roi_y

    # bounds checking
    H, W = images_use.shape[1], images_use.shape[2]
    if not (0 <= x0 < x1 <= W and 0 <= y0 < y1 <= H):
        raise ValueError(f"ROI out of bounds for image shape (H,W)=({H},{W}). "
                         f"Got roi_x={roi_x}, roi_y={roi_y}")

    # --- collect ROI pixels across many images ---
    # shape: (n_images, roi_h, roi_w) -> flatten to 1D
    roi_cube = images_use[idx, y0:y1, x0:x1]
    roi_vals = roi_cube.reshape(-1)

    # drop non-finite just in case
    roi_vals = roi_vals[np.isfinite(roi_vals)]
    if roi_vals.size == 0:
        raise ValueError("ROI contains no finite values.")

    # --- stats ---
    mean = float(np.mean(roi_vals))
    median = float(np.median(roi_vals))
    std = float(np.std(roi_vals, ddof=1)) if roi_vals.size > 1 else 0.0

    # robust sigma from MAD (median absolute deviation)
    mad = float(np.median(np.abs(roi_vals - median))) if robust else None
    sigma_mad = float(1.4826 * mad) if robust else None

    p01, p05, p95, p99 = [float(x) for x in np.percentile(roi_vals, [1, 5, 95, 99])]

    stats = dict(
        target_timestamp=int(target_timestamp),
        bk_subtract=bool(bk_subtract),
        bk_timestamp=int(bk_timestamp) if bk_subtract else None,
        roi_x=tuple(roi_x),
        roi_y=tuple(roi_y),
        n_images_total=int(n_total),
        n_images_used=int(len(idx)),
        n_roi_pixels=int(roi_vals.size),
        mean=mean,
        median=median,
        std=std,
        mad=mad,
        sigma_mad=sigma_mad,
        p01=p01,
        p05=p05,
        p95=p95,
        p99=p99,
    )

    if verbose:
        sig = sigma_mad if robust else std
        sig_name = "sigma_mad" if robust else "std"
        print(
            f"[bg stats] ts={target_timestamp} | bk={bk_subtract} "
            f"| ROI x={roi_x} y={roi_y} | imgs used={len(idx)}\n"
            f"  median={median:.4g}, mean={mean:.4g}, std={std:.4g}, "
            f"{sig_name}={sig:.4g}, p95={p95:.4g}, p99={p99:.4g}"
        )

    if return_samples:
        stats["samples"] = roi_vals  # can be big; only enable when debugging

    return stats
```

```python
def analysis_plots2(exclude_list, proj_list, x_axis, save_name=None, save=False, upper_limit=None,
                    show_individual=True, title_pre="Mean ± STD (pre-alignment)",
                    title_post="Mean ± STD (after peak alignment)", x_lim=[-200,200], axis_ps_fs="fs"):
    if axis_ps_fs == "ps":
        scale = 1000
        x_axis_label = "Time (ps)"
    else:
        scale = 1
        x_axis_label = "Time (fs)"

    proj_array = np.asarray(proj_list, dtype=float)
    N, W = proj_array.shape
    assert len(x_axis) == W, "x_axis length must match projection length"

    cand_idx = np.arange(min(N, upper_limit)) if upper_limit is not None else np.arange(N)
    excl = np.unique([i for i in exclude_list if 0 <= i < N])
    keep_idx = np.setdiff1d(cand_idx, excl)
    kept = proj_array[keep_idx]
    if kept.size == 0:
        print("No shots to analyze after exclusions.")
        return None

    x = np.asarray(x_axis)
    ddof = 1 if kept.shape[0] > 1 else 0
    mean_pre = np.nanmean(kept, axis=0)
    std_pre  = np.nanstd(kept, axis=0, ddof=ddof)

    fig_pre, axs_pre = plt.subplots(1, 2, figsize=(12, 4), sharex=True)
    if show_individual:
        for row in kept:
            axs_pre[0].plot(x/scale, row, lw=0.5, alpha=0.5)
    axs_pre[0].set_title("Individual profiles (pre-alignment)")
    axs_pre[0].set_xlabel(x_axis_label)
    axs_pre[0].set_ylabel("Current (A)")
    axs_pre[0].set_xlim(x_lim[0], x_lim[1])

    axs_pre[1].plot(x/scale, mean_pre, lw=2, label='Mean (pre)')
    axs_pre[1].fill_between(x/scale, mean_pre - std_pre, mean_pre + std_pre, alpha=0.35, label='±1 STD')
    axs_pre[1].set_title(title_pre)
    axs_pre[1].set_xlabel(x_axis_label)
    axs_pre[1].set_ylabel("Current (A)")
    axs_pre[1].set_xlim(x_lim[0], x_lim[1])
    axs_pre[1].legend()
    fig_pre.tight_layout()
    if save and save_name:
        fig_pre.savefig(f"{save_name}_pre.pdf", dpi=300)

    plt.show()

    return {
        "keep_idx": keep_idx,
        "mean_pre": mean_pre, "std_pre": std_pre,
        "kept": kept, "x": x
    }
```

```python
def plot_xtcav_four(target_timestamp, trex_stamp, indices,
                    save_name="xtcav_four",
                    hard_code_streak=7.37, image_summation_threshold=1000,
                    moving_average_window=30, y_moving_average_window=None, threshold=5,
                    bk_subtract=False, bk_timestamp=None, color='jet',
                    y_lim=[-30,30], x_lim=[-200,200], axis_fs_ps="fs", resolution_hardcoded=None):

    assert len(indices) == 4, "Please pass exactly four indices."

    # --- Load TREX metadata ---
    filtered_entries = trex_struct_sorted[trex_struct_sorted['timestamp'] == trex_stamp]
    filename = filtered_entries['filename'][0]
    trex = beam_time_dir + filename

    if y_moving_average_window is None:
        y_moving_average_window = moving_average_window

    if axis_fs_ps == "ps":
        scale = 1000
        x_axis_label = "Time (ps)"
    else:
        scale = 1
        x_axis_label = "Time (fs)"

    with h5py.File(trex, "r") as f:
        dispersion  = np.array(f["data/dispersion"]).item()
        mean_energy = np.array(f["data/mean_erg"]).item()
        px2um       = np.array(f["data/px2um"]).item()

    if resolution_hardcoded is not None:
        px2um = resolution_hardcoded

    xtcav_params = {
        "resolution": px2um,
        "dispersion": dispersion*1e-6*-1,
        "streak": hard_code_streak,
        "energy": mean_energy,
        "cavity_frequency": 11.424e9,
    }

    populate_correlation_entries(corr_struct_sorted, beam_time_dir, target_timestamp)
    charges, images, valid_charge, valid_image = get_charge_and_image_by_timestamp(corr_struct_sorted, target_timestamp)

    if bk_subtract and (bk_timestamp is not None):
        populate_correlation_entries(corr_struct_sorted, beam_time_dir, bk_timestamp)
        bk_charges, bk_images, bk_valid_charge, bk_valid_image = get_charge_and_image_by_timestamp(corr_struct_sorted, bk_timestamp)
        bk_mean = np.mean(bk_images, axis=0)
        images = images - bk_mean

    px2fs, px2mev, xticks, yticks = get_tcavAxes_conversion_ticks(
        xtcav_params["energy"], xtcav_params["streak"], xtcav_params["dispersion"],
        xtcav_params["cavity_frequency"], xtcav_params["resolution"], 3e8
    )
    print("Axis info")
    print("px2fs")
    print(px2fs)
    print("px2mev")
    print(px2mev)

    def process_one(i):
        if i < 0 or i >= images.shape[0]:
            print("Here")
            print(images.shape[0])
            return None, None, None, None, None, f"Index {i} out of range."

        filtered = process_tcav_image_v1(images[i, :, :], threshold=threshold)
        img = filtered.T
        h, w = img.shape
        img_sum = np.sum(img)
        print(img_sum)
        if img_sum < image_summation_threshold:
            
            return None, None, None, None, None, f"Skipping index {i}: sum<{image_summation_threshold}"

        charge = float(charges[i])
        img_scaled = (img / img_sum)

        scaler_xproj = np.abs(charge * 1e-12 / (np.sum(img_scaled) * px2fs * 1e-15)) # Charge in pc over time step in fs
        scaler_yproj = np.abs(charge * 1e-12 / (np.sum(img_scaled) * px2mev * 1e6))

        # centroid recenter (integer)
        x_idx = np.arange(w)
        y_idx = np.arange(h)
        total = img_scaled.sum()
        x_c = (img_scaled.sum(axis=0) @ x_idx) / total
        y_c = (img_scaled.sum(axis=1) @ y_idx) / total
        dx = int(round(w/2 - x_c))
        dy = int(round(h/2 - y_c))
        img_centered = np.roll(np.roll(img_scaled, shift=dy, axis=0), shift=dx, axis=1)

        # physical axes (centered)
        x_phys = (np.arange(w) - w/2) * px2fs
        y_phys = (np.arange(h) - h/2) * px2mev

        # projections + smoothing + padding
        x_projection = np.sum(img_centered*scaler_xproj, axis=0)
        xp = moving_average(x_projection, moving_average_window)
        pad = (x_projection.shape[0] - xp.shape[0]) // 2
        xp = np.pad(xp, (pad, x_projection.shape[0]-xp.shape[0]-pad), mode='edge')

        y_projection = np.sum(img_centered*scaler_yproj, axis=1)
        yp = moving_average(y_projection, y_moving_average_window)
        pad = (y_projection.shape[0] - yp.shape[0]) // 2
        yp = np.pad(yp, (pad, y_projection.shape[0]-yp.shape[0]-pad), mode='edge')


        return img_centered, x_phys, y_phys, (xp, yp), charge, None

    processed = []
    plotted = []
    for i in indices:
        processed.append((i, *process_one(i)))

    fig, axes = plt.subplots(2, 2, figsize=(12, 9), constrained_layout=True)
    axes = axes.ravel()

    for ax, (i, img_centered, x_phys, y_phys, projs, charge_i, err) in zip(axes, processed):
        if charge_i is not None:
            ax.set_title(f"idx {i} | charge={charge_i:.3f}")
        else:
            ax.set_title(f"idx {i}")

        if err is not None or img_centered is None:
            ax.text(0.5, 0.5, err if err else "No data", ha='center', va='center', transform=ax.transAxes)
            ax.axis('off')
            continue
        print("Image shape")
        print(img_centered.shape)
        img_centered_maxNormalized = img_centered/(np.max(img_centered))
        # im = ax.imshow(img_centered, cmap=color, origin='upper',
        #                extent=[x_phys[0]/scale, x_phys[-1]/scale, y_phys[-1], y_phys[0]],
        #                aspect='auto', vmin=0)   # <-- no shared vmax
        im = ax.imshow(img_centered_maxNormalized, cmap=color, origin='upper',
                       extent=[x_phys[0]/scale, x_phys[-1]/scale, y_phys[-1], y_phys[0]],
                       aspect='auto', vmin=0)   # <-- no shared vmax
        ax.axhline(0, ls='--', color="black", lw=0.8)
        ax.axvline(0, ls='--', color="black", lw=0.8)
        ax.set_xlim(x_lim[0], x_lim[1])
        ax.set_ylim(y_lim[0], y_lim[1])
        ax.set_xlabel(x_axis_label)
        ax.set_ylabel("ΔEnergy (MeV)")

        # per-panel colorbar (key requested change)
        cbar = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.02)
        cbar.set_label("Intensity (a.u.)", fontsize=9)

        # overlay x-projection on right axis
        xp, yp = projs
        axr = ax.twinx()
        axr.plot(x_phys/scale, xp, color='red', lw=0.6, alpha=0.95)
        axr.set_ylabel("Current (A)", color='red')
        axr.tick_params(axis='y', colors='red', labelsize=8)
        axr.spines['right'].set_edgecolor('red')
        axr.set_xlim(ax.get_xlim())
        axr.set_ylim(0, 1.1 * max(1e-12, xp.max()))

        # # top axis for y projection
        # axt = ax.twiny()
        # axt.plot(yp, y_phys, color='red', lw=0.6, alpha=0.95)
        # axt.set_xlabel("Y proj (a.u.)", color='red')
        # axt.tick_params(axis='x', colors='red', labelsize=8)
        # axt.spines['top'].set_edgecolor('red')
        # axt.set_ylim(ax.get_ylim())
        # axt.set_xlim(0, 1.1 * max(1e-12, yp.max()))

        # top axis for y projection
        axt = ax.twiny()
        axt.plot(yp/np.max(yp), y_phys, color='red', lw=0.6, alpha=0.95)
        axt.set_xlabel("Y proj (a.u.)", color='red')
        axt.tick_params(axis='x', colors='red', labelsize=8)
        axt.spines['top'].set_edgecolor('red')
        # axt.set_ylim(ax.get_ylim())
        # axt.set_xlim(0, 1.1)
        plotted.append(img_centered_maxNormalized)

    if save_name:
        fig.savefig(f"{save_name}.pdf", dpi=300)
    plt.show()

    return {"indices": indices, "processed": processed, "px_scale": (px2fs, px2mev)}, plotted
    
def plot_xtcav_four_proj(
    target_timestamp,
    trex_stamp,
    proj_indices,
    index_map,
    **kwargs
):
    """
    proj_indices: indices into proj_list (i.e., the order produced by prep_xtcav)
    index_map: output from prep_xtcav (contains 'proj_to_raw')
    kwargs: forwarded to plot_xtcav_four (thresholds, bk_subtract, etc.)
    """
    assert len(proj_indices) == 4, "Pass exactly four projection indices."
    proj_to_raw = index_map["proj_to_raw"]

    raw_indices = []
    for k in proj_indices:
        if k not in proj_to_raw:
            raise IndexError(f"Projection index {k} not found. Valid range is 0..{len(proj_to_raw)-1}.")
        raw_indices.append(int(proj_to_raw[k]))

    # call your existing function with raw indices
    return plot_xtcav_four(
        target_timestamp=target_timestamp,
        trex_stamp=trex_stamp,
        indices=raw_indices,
        **kwargs
    )
```

```python
def pick_random_indices(low, high, n=4, exclude=None, seed=None):
    if exclude is None:
        exclude = []

    if seed is not None:
        random.seed(seed)

    # Build allowed pool
    allowed = [i for i in range(low, high + 1) if i not in exclude]

    if len(allowed) < n:
        raise ValueError("Not enough valid numbers to sample from.")

    return random.sample(allowed, n)
```

```python
def print_h5_structure(g, indent=0):
    for key in g:
        item = g[key]
        print("  " * indent + f"{key} ({type(item)})")
        if isinstance(item, h5py.Group):
            print_h5_structure(item, indent + 1)

def plot_single_xtcav(
    image, charge, trex_stamp,
    save_name=None,
    hard_code_streak=7.37, image_summation_threshold=1000,
    moving_average_window=30, y_moving_average_window=None, threshold=5,
    bk_subtract=False, bk_timestamp=None, color='jet',
    y_lim=(-30, 30), x_lim=(-200, 200), axis_fs_ps="fs",
    resolution_hardcoded=None,
):
    # --- Load TREX metadata ---
    filtered_entries = trex_struct_sorted[trex_struct_sorted['timestamp'] == trex_stamp]

    if y_moving_average_window is None:
        y_moving_average_window = moving_average_window

    if axis_fs_ps == "ps":
        scale = 1000.0
        x_axis_label = "Time (ps)"
    else:
        scale = 1.0
        x_axis_label = "Time (fs)"

    with h5py.File(trex, "r") as f:
        dispersion  = np.array(f["data/dispersion"]).item()
        mean_energy = np.array(f["data/mean_erg"]).item()
        px2um       = np.array(f["data/px2um"]).item()

    if resolution_hardcoded is not None:
        px2um = resolution_hardcoded

    xtcav_params = {
        "resolution": px2um,
        "dispersion": dispersion * 1e-6 * -1,
        "streak": hard_code_streak,
        "energy": mean_energy,
        "cavity_frequency": 11.424e9,
    }

    # --- optional background subtraction ---
    if bk_subtract and (bk_timestamp is not None):
        populate_correlation_entries(corr_struct_sorted, beam_time_dir, bk_timestamp)
        bk_charges, bk_images, bk_valid_charge, bk_valid_image = get_charge_and_image_by_timestamp(
            corr_struct_sorted, bk_timestamp
        )
        bk_mean = np.mean(bk_images, axis=0)
        image = image - bk_mean

    # --- axis conversion ---
    px2fs, px2mev, xticks, yticks = get_tcavAxes_conversion_ticks(
        xtcav_params["energy"], xtcav_params["streak"], xtcav_params["dispersion"],
        xtcav_params["cavity_frequency"], xtcav_params["resolution"], 3e8
    )

    # --- preprocess image ---
    filtered = process_tcav_image_v1(image[:, :], threshold=threshold)
    img = filtered.T  # keep your convention
    h, w = img.shape
    img_sum = np.sum(img)
    print(img_sum)
    print("here")
    if img_sum < image_summation_threshold:
        # still show something minimal (optional)
        fig, ax = plt.subplots(figsize=(3, 3), constrained_layout=True)
        ax.text(0.5, 0.5, f"sum<{image_summation_threshold}\n({img_sum:.1f})",
                ha="center", va="center", transform=ax.transAxes)
        ax.axis("off")
        plt.show()
        return {"processed": None, "px_scale": (px2fs, px2mev), "error": "sum_below_threshold"}

    img_scaled = img / img_sum

    scaler_xproj = np.abs(charge * 1e-12 / (np.sum(img_scaled) * px2fs * 1e-15))
    scaler_yproj = np.abs(charge * 1e-12 / (np.sum(img_scaled) * px2mev * 1e6))

    # --- centroid recenter (integer) ---
    x_idx = np.arange(w)
    y_idx = np.arange(h)
    total = img_scaled.sum()
    x_c = (img_scaled.sum(axis=0) @ x_idx) / total
    y_c = (img_scaled.sum(axis=1) @ y_idx) / total
    dx = int(round(w / 2 - x_c))
    dy = int(round(h / 2 - y_c))
    img_centered = np.roll(np.roll(img_scaled, shift=dy, axis=0), shift=dx, axis=1)

    # --- physical axes (centered) ---
    x_phys = (np.arange(w) - w / 2) * px2fs
    y_phys = (np.arange(h) - h / 2) * px2mev

    # --- projections + smoothing + padding ---
    x_projection = np.sum(img_centered * scaler_xproj, axis=0)
    xp = moving_average(x_projection, moving_average_window)
    pad = (x_projection.shape[0] - xp.shape[0]) // 2
    xp = np.pad(xp, (pad, x_projection.shape[0] - xp.shape[0] - pad), mode="edge")

    y_projection = np.sum(img_centered * scaler_yproj, axis=1)
    yp = moving_average(y_projection, y_moving_average_window)
    pad = (y_projection.shape[0] - yp.shape[0]) // 2
    yp = np.pad(yp, (pad, y_projection.shape[0] - yp.shape[0] - pad), mode="edge")

    # --- plotting (single axes) ---
    fig, ax = plt.subplots(figsize=(3, 3), constrained_layout=True)
    ax.set_title(f"charge={charge:.3f}")

    img_norm = img_centered / np.max(img_centered)
    im = ax.imshow(
        img_norm,
        cmap=color,
        origin="upper",
        extent=[x_phys[0] / scale, x_phys[-1] / scale, y_phys[-1], y_phys[0]],
        aspect="auto",
        vmin=0,
    )

    ax.axhline(0, ls="--", color="black", lw=0.8)
    ax.axvline(0, ls="--", color="black", lw=0.8)
    ax.set_xlim(*x_lim)
    ax.set_ylim(*y_lim)
    ax.set_xlabel(x_axis_label)
    ax.set_ylabel("ΔEnergy (MeV)")

    cbar = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.02)
    cbar.set_label("Intensity (a.u.)", fontsize=9)

    axr = ax.twinx()
    axr.plot(x_phys / scale, xp, color="red", lw=0.6, alpha=0.95)
    axr.set_ylabel("Current (A)", color="red")
    axr.tick_params(axis="y", colors="red", labelsize=8)
    axr.spines["right"].set_edgecolor("red")
    axr.set_xlim(ax.get_xlim())
    axr.set_ylim(0, 1.1 * max(1e-12, xp.max()))

    axt = ax.twiny()
    axt.plot(yp / np.max(yp), y_phys, color="red", lw=0.6, alpha=0.95)
    axt.set_xlabel("Y proj (a.u.)", color="red")
    axt.tick_params(axis="x", colors="red", labelsize=8)
    axt.spines["top"].set_edgecolor("red")

    if save_name:
        fig.savefig(f"{save_name}.pdf", dpi=300, bbox_inches="tight")

    plt.show()

    return {
        "processed": (img_centered, x_phys, y_phys, (xp, yp), charge),
        "px_scale": (px2fs, px2mev),
        "plotted":img_norm
    }
```

```python
def process_single_xtcav_with_moments(
    image,
    charge_pc,
    trex_file_path,
    hard_code_streak=7.37,
    image_summation_threshold=1.0,
    moving_average_window=30,
    y_moving_average_window=None,
    threshold=5,
    bk_subtract=False,
    bk_timestamp=None,
    resolution_hardcoded=None,
    axis_fs_ps="fs",
):
    """
    Returns:
      img2d_use : (H,W) processed image (normalized by total image sum)
      x_phys    : fs
      y_phys    : MeV
      E_com     : MeV
      sigmaE2   : MeV^2
      sigmaE    : MeV
      I         : A
      img_norm  : for display only
    """
    if y_moving_average_window is None:
        y_moving_average_window = moving_average_window

    scale = 1000.0 if axis_fs_ps == "ps" else 1.0

    import h5py

    with h5py.File(trex_file_path, "r") as f:
        dispersion  = np.array(f["data/dispersion"]).item()
        mean_energy = np.array(f["data/mean_erg"]).item()
        px2um       = np.array(f["data/px2um"]).item()

    if resolution_hardcoded is not None:
        px2um = resolution_hardcoded

    xtcav_params = {
        "resolution": px2um,
        "dispersion": dispersion * 1e-6 * -1,
        "streak": hard_code_streak,
        "energy": mean_energy,
        "cavity_frequency": 11.424e9,
    }

    img_in = np.asarray(image, dtype=float)
    if bk_subtract and (bk_timestamp is not None):
        populate_correlation_entries(corr_struct_sorted, beam_time_dir, bk_timestamp)
        bk_charges, bk_images, *_ = get_charge_and_image_by_timestamp(corr_struct_sorted, bk_timestamp)
        img_in = img_in - np.mean(bk_images, axis=0)

    px2fs, px2mev, *_ = get_tcavAxes_conversion_ticks(
        xtcav_params["energy"],
        xtcav_params["streak"],
        xtcav_params["dispersion"],
        xtcav_params["cavity_frequency"],
        xtcav_params["resolution"],
        3e8,
    )

    filtered = process_tcav_image_v1(img_in[:, :], threshold=threshold)

    img = np.asarray(filtered.T, dtype=float)
    img = np.maximum(img, 0.0)

    h, w = img.shape
    img_sum = np.sum(img)
    if img_sum < image_summation_threshold:
        return {"ok": False, "error": "sum_below_threshold", "img_sum": float(img_sum)}

    # normalized image for moment calculations
    img_scaled = img / img_sum

    x_phys = (np.arange(w) - w / 2) * px2fs
    y_phys = (np.arange(h) - h / 2) * px2mev

    # current profile
    scaler_xproj = np.abs(charge_pc * 1e-12 / (np.sum(img_scaled) * px2fs * 1e-15))
    x_projection = np.sum(img_scaled * scaler_xproj, axis=0)
    xp = moving_average(x_projection, moving_average_window)
    pad = (x_projection.shape[0] - xp.shape[0]) // 2
    xp = np.pad(xp, (pad, x_projection.shape[0] - xp.shape[0] - pad), mode="edge")
    I = _nanfill_1d(xp)

    # first moment
    EPS_COL = 1e-15
    col_sum = img_scaled.sum(axis=0)
    col_sum = np.where(col_sum < EPS_COL, np.nan, col_sum)
    E_com = (img_scaled.T @ y_phys) / col_sum
    E_com = _nanfill_1d(E_com)

    # second central moment
    dy = y_phys[:, None] - E_com[None, :]
    sigmaE2 = np.sum(img_scaled * (dy ** 2), axis=0) / col_sum
    sigmaE2 = _nanfill_1d(sigmaE2)
    sigmaE2 = np.maximum(sigmaE2, 0.0)
    sigmaE = np.sqrt(sigmaE2)

    img_norm = img_scaled / max(1e-12, np.max(img_scaled))

    return {
        "ok": True,
        "img2d_use": img_scaled,
        "img_norm": img_norm,
        "x_phys": x_phys / scale,
        "y_phys": y_phys,
        "E_com": E_com,
        "sigmaE2": sigmaE2,
        "sigmaE": sigmaE,
        "I": I,
        "xp": xp,
        "px2fs": px2fs,
        "px2mev": px2mev,
    }

def recompute_moments_from_img(img2d_use, x_fs, y_mev, charge_pc, moving_average_window):
    """
    Recompute I(t), E_com(t), sigmaE2(t), sigmaE(t) after a time shift is applied to the 2D image.
    """
    xproj = np.sum(img2d_use, axis=0)
    xproj_s = moving_average(xproj, moving_average_window)
    pad = (xproj.shape[0] - xproj_s.shape[0]) // 2
    xproj_s = np.pad(xproj_s, (pad, xproj.shape[0] - xproj_s.shape[0] - pad), mode="edge")

    dt_s = float(np.median(np.diff(x_fs)) * 1e-15)
    Q_C = float(charge_pc) * 1e-12
    I_A = xproj_s * (Q_C / max(1e-30, np.sum(xproj_s) * dt_s))
    I_A = _nanfill_1d(I_A)

    col_sum = np.sum(img2d_use, axis=0)
    col_sum = np.where(col_sum < 1e-18, np.nan, col_sum)

    Ecom = (img2d_use.T @ y_mev) / col_sum
    Ecom = _nanfill_1d(Ecom)

    dy = y_mev[:, None] - Ecom[None, :]
    sigmaE2 = np.sum(img2d_use * (dy ** 2), axis=0) / col_sum
    sigmaE2 = _nanfill_1d(sigmaE2)
    sigmaE2 = np.maximum(sigmaE2, 0.0)
    sigmaE = np.sqrt(sigmaE2)

    return I_A, Ecom, sigmaE2, sigmaE


# ============================================================
# Reconstruction engines
# ============================================================
def reconstruct_delta_offset_corrected(
    x_fs,
    E_on,
    E_off,
    I_on,
    mask,
    target_J,
):
    """
    DELTA-method with additive offset correction following the SLAC slide.

    Raw:
      P_raw(t) = [E_off(t) - E_on(t)] * I_on(t) / 1000   [GW]

    Compute missing energy:
      U_offset = U_GDET - ∫ P_raw(t) dt

    Convert to an energy-per-electron shift:
      dE_shift_MeV = U_offset / (Q_window * 1e6)

    Correct the ON curve:
      E_on_corr(t) = E_on(t) - dE_shift_MeV

    Recompute:
      P_corr(t) = [E_off(t) - E_on_corr(t)] * I_on(t) / 1000
    """
    x_fs = np.asarray(x_fs, dtype=float)
    E_on = np.asarray(E_on, dtype=float)
    E_off = np.asarray(E_off, dtype=float)
    I_on = np.asarray(I_on, dtype=float)
    mask = np.asarray(mask, dtype=bool)

    # raw delta method
    dE_raw = E_off - E_on                       # MeV
    P_raw_GW = dE_raw * I_on / 1000.0          # GW

    # integrated raw reconstructed X-ray energy
    U_raw_J = integrate_power_J(x_fs, P_raw_GW, mask=mask)

    # integrated beam charge over reconstruction window
    Q_window_C = integrate_current_C(x_fs, I_on, mask=mask)

    # missing energy in Joules
    if not np.isfinite(U_raw_J) or not np.isfinite(Q_window_C) or Q_window_C <= 0 or target_J is None:
        U_offset_J = np.nan
        dE_shift_MeV = np.nan
        E_on_corr = E_on.copy()
    else:
        U_offset_J = target_J - U_raw_J

        # convert Joules/electron -> eV, then -> MeV
        # since 1 eV = 1 J/C for one electron charge basis, divide by charge in Coulombs
        dE_shift_MeV = U_offset_J / Q_window_C / 1e6

        # shift ON curve exactly as in slide
        E_on_corr = E_on - dE_shift_MeV

    # recompute corrected delta method
    dE_corr = E_off - E_on_corr
    P_corr_GW = dE_corr * I_on / 1000.0
    U_corr_J = integrate_power_J(x_fs, P_corr_GW, mask=mask)

    return {
        "E_on_corr_MeV": E_on_corr,
        "dE_raw_MeV": dE_raw,
        "P_raw_GW": P_raw_GW,
        "U_raw_J": U_raw_J,
        "Q_window_C": Q_window_C,
        "U_offset_J": U_offset_J,
        "dE_shift_MeV": dE_shift_MeV,
        "dE_corr_MeV": dE_corr,
        "P_corr_GW": P_corr_GW,
        "U_corr_J": U_corr_J,
    }

def reconstruct_sigma_normalized(
    x_fs,
    sigmaE2_on,
    sigmaE2_off,
    I_on,
    mask,
    target_J,
):
    """
    SIGMA-method:
      P_sigma(t) ∝ [sigma_on^2(t) - sigma_off^2(t)] * I_on(t)^(2/3)
    then multiplicatively normalize to target_J over mask.
    """
    x_fs = np.asarray(x_fs, dtype=float)
    sigmaE2_on = np.asarray(sigmaE2_on, dtype=float)
    sigmaE2_off = np.asarray(sigmaE2_off, dtype=float)
    I_on = np.asarray(I_on, dtype=float)
    mask = np.asarray(mask, dtype=bool)

    dsig2 = sigmaE2_on - sigmaE2_off
    Iterm = np.power(np.maximum(I_on, 0.0), 2.0 / 3.0)
    P_shape = dsig2 * Iterm

    P_norm, U_raw_J, scale = normalize_profile_to_energy(
        t_fs=x_fs,
        P_arb=P_shape,
        mask=mask,
        target_J=target_J,
    )

    return {
        "dsig2_MeV2": dsig2,
        "P_shape_arb": P_shape,
        "P_norm": P_norm,
        "U_raw_J": U_raw_J,
        "scale": scale,
    }
# ============================================================
# Plot helpers
# ============================================================
def xtcav_plot_from_processed(proc, title=""):
    x = np.asarray(proc["x_phys"], dtype=float)
    y = np.asarray(proc["y_phys"], dtype=float)
    img = np.asarray(proc["img_norm"], dtype=float)

    y_plot = y.copy()
    if PLOT_RECENTER_Y_FOR_DISPLAY:
        if RE_CENTER_Y_BY == "median":
            row_w = np.sum(img, axis=1)
            row_w = row_w / max(1e-12, np.sum(row_w))
            y0 = np.sum(y * row_w)
        elif RE_CENTER_Y_BY == "com":
            w = np.sum(img)
            y0 = np.sum(img * y[:, None]) / max(1e-12, w)
        else:
            raise ValueError("RE_CENTER_Y_BY must be 'median' or 'com'")
        y_plot = y_plot - y0

    fig, ax = plt.subplots(figsize=(3.0, 3.0), constrained_layout=True)
    ax.set_title(title)
    im = ax.imshow(
        img, cmap=CMAP, origin="upper",
        extent=[x[0], x[-1], y_plot[-1], y_plot[0]],
        aspect="auto", vmin=0
    )
    ax.set_xlim(*XLIM_FS)
    ax.set_ylim(*IMAGE_YLIM_MEV)
    ax.set_xlabel("Time (fs)")
    ax.set_ylabel("ΔEnergy (MeV)")
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.02)
    plt.show()
    return fig

def plot_alignment_fom(sb_grid, rhoI_curve, rhoE_curve, score_curve, sb_best, idx_best, x_ref):
    fig, ax = plt.subplots(figsize=(7.6, 4.2), constrained_layout=True)
    ax.set_title(f"Alignment FoM vs integer time-shift for best conservative OFF shot idx={idx_best}")
    ax.set_xlabel(f"Shift (bins)   [dt={np.median(np.diff(x_ref)):.2f} fs/bin ; + = OFF delayed]")
    ax.set_ylabel("Correlation (0–1)")
    ax.plot(sb_grid, rhoI_curve, lw=1.8, label="rhoI (current)")
    ax.plot(sb_grid, rhoE_curve, lw=1.4, ls="--", label="rhoE (COM energy)")
    ax.plot(sb_grid, score_curve, lw=1.4, ls=":", label="combined score")
    ax.axvline(0, lw=0.8, alpha=0.5)
    ax.axvline(sb_best, lw=1.0, alpha=0.5, label=f"best current shift = {sb_best:+d}")
    ax.legend(fontsize=9)
    plt.show()
    return fig

def plot_deltaE_panels(
    img2d_off0,
    y_mev_off,
    x_ref,
    E_on,
    charge_off_pc,
    moving_avg_window,
    mwin,
    sb_best,
    offset_bins_list,
    idx_best,
    rhoI_best,
    rhoE_best,
):
    fig, axes = plt.subplots(2, 3, figsize=(12.5, 6.5), constrained_layout=True, sharex=True, sharey=True)
    axes = axes.ravel()

    for ax, offb in zip(axes, offset_bins_list):
        sb = int(sb_best + offb)
        img2d_sh = shift_2d_time_zerofill(img2d_off0, sb)

        _, E_off_s, _, _ = recompute_moments_from_img(
            img2d_use=img2d_sh,
            x_fs=x_ref,
            y_mev=y_mev_off,
            charge_pc=charge_off_pc,
            moving_average_window=moving_avg_window,
        )

        dE = E_off_s - E_on
        ax.plot(x_ref, dE, lw=1.8)
        ax.axhline(0, ls="--", lw=0.8, alpha=0.7)
        if mwin.sum() > 1:
            ax.axvspan(x_ref[mwin][0], x_ref[mwin][-1], alpha=0.08)
        ax.set_title(f"shift = {sb:+d} bins  (best {sb_best:+d} + {offb:+d})")
        ax.set_xlim(*XLIM_FS)

    for ax in axes[3:]:
        ax.set_xlabel("Time (fs)")
    for ax in axes[::3]:
        ax.set_ylabel("Δ⟨E⟩ (MeV)")

    fig.suptitle(
        f"Δ⟨E⟩(t) = ⟨E_off⟩ - ⟨E_on⟩ for best OFF idx={idx_best} (current-aligned)\n"
        f"Best shift = {sb_best:+d} bins ({sb_best*np.median(np.diff(x_ref)):+.1f} fs), "
        f"rhoI={rhoI_best:.3f}, rhoE={rhoE_best:.3f}",
        y=1.02
    )
    plt.show()
    return fig
```

```python


# ============================================================
# Helpers
# ============================================================
def maybe_savefig(fig, name_pdf):
    if not SAVE or (SAVE_DIR is None):
        return
    os.makedirs(SAVE_DIR, exist_ok=True)
    if not name_pdf.lower().endswith(".pdf"):
        name_pdf += ".pdf"
    path = os.path.join(SAVE_DIR, name_pdf)
    fig.savefig(path, dpi=300, bbox_inches="tight", format="pdf")
    print(f"Saved: {path}")

def charge_to_pC(q):
    q = np.asarray(q, dtype=float)
    if CHARGE_UNITS.lower() == "pc":
        return q
    if CHARGE_UNITS.lower() == "c":
        return q * 1e12
    raise ValueError("CHARGE_UNITS must be 'pC' or 'C'")

def _get_stamp(stamps, i):
    return stamps if np.isscalar(stamps) else stamps[i]

def _interp_to(x_src, y_src, x_tgt):
    x_src = np.asarray(x_src, dtype=float)
    y_src = np.asarray(y_src, dtype=float)
    m = np.isfinite(x_src) & np.isfinite(y_src)
    if m.sum() < 3:
        return np.full_like(x_tgt, np.nan, dtype=float)
    return np.interp(x_tgt, x_src[m], y_src[m], left=np.nan, right=np.nan)

def _nanfill_1d(y):
    y = np.asarray(y, dtype=float)
    if np.all(np.isfinite(y)):
        return y
    x = np.arange(len(y))
    m = np.isfinite(y)
    if m.sum() < 3:
        return np.nan_to_num(y, nan=0.0)
    y2 = y.copy()
    y2[~m] = np.interp(x[~m], x[m], y[m])
    return y2

def _normalize(y):
    y = _nanfill_1d(np.asarray(y, dtype=float))
    mu = np.mean(y)
    sig = np.std(y)
    if sig < 1e-12:
        return y * 0.0
    return (y - mu) / sig

def corr01(a, b):
    a = _normalize(a)
    b = _normalize(b)
    c = np.corrcoef(a, b)[0, 1]
    if not np.isfinite(c):
        return 0.0
    return 0.5 * (c + 1.0)

def shift_2d_time_zerofill(img2d, shift_bins):
    """
    Shift processed 2D image in TIME (axis=1), no wrap-around.
    +shift => delayed => move content right.
    """
    img = np.asarray(img2d, dtype=float)
    out = np.zeros_like(img, dtype=float)
    s = int(shift_bins)
    if s == 0:
        out[:] = img
    elif s > 0:
        out[:, s:] = img[:, :-s]
    else:
        out[:, :s] = img[:, -s:]
    return out

def integrate_power_J(t_fs, P_GW, mask=None):
    t_s = np.asarray(t_fs, dtype=float) * 1e-15
    P_W = np.asarray(P_GW, dtype=float) * 1e9
    if mask is None:
        mask = np.ones_like(P_W, dtype=bool)
    mask = np.asarray(mask, dtype=bool)
    if mask.sum() < 2:
        return np.nan
    return np.trapz(P_W[mask], t_s[mask])

def integrate_current_C(t_fs, I_A, mask=None):
    t_s = np.asarray(t_fs, dtype=float) * 1e-15
    I_A = np.asarray(I_A, dtype=float)
    if mask is None:
        mask = np.ones_like(I_A, dtype=bool)
    mask = np.asarray(mask, dtype=bool)
    if mask.sum() < 2:
        return np.nan
    return np.trapz(I_A[mask], t_s[mask])

def normalize_profile_to_energy(t_fs, P_arb, mask, target_J):
    """
    Multiplicative normalization for sigma-method.
    """
    P_arb = np.asarray(P_arb, dtype=float)
    E_raw = integrate_power_J(t_fs, P_arb, mask=mask)
    if not np.isfinite(E_raw) or E_raw <= 0 or target_J is None:
        return P_arb, E_raw, np.nan
    scale = target_J / E_raw
    return P_arb * scale, E_raw, scale


# ============================================================
# Core processing
# ============================================================
def process_single_xtcav_with_moments(
    image,
    charge_pc,
    trex_file_path,
    hard_code_streak=7.37,
    image_summation_threshold=1.0,
    moving_average_window=30,
    y_moving_average_window=None,
    threshold=5,
    bk_subtract=False,
    bk_timestamp=None,
    resolution_hardcoded=None,
    axis_fs_ps="fs",
):
    """
    Returns:
      img2d_use : (H,W) processed image (normalized by total image sum)
      x_phys    : fs
      y_phys    : MeV
      E_com     : MeV
      sigmaE2   : MeV^2
      sigmaE    : MeV
      I         : A
      img_norm  : for display only
    """
    if y_moving_average_window is None:
        y_moving_average_window = moving_average_window

    scale = 1000.0 if axis_fs_ps == "ps" else 1.0

    import h5py

    with h5py.File(trex_file_path, "r") as f:
        dispersion  = np.array(f["data/dispersion"]).item()
        mean_energy = np.array(f["data/mean_erg"]).item()
        px2um       = np.array(f["data/px2um"]).item()

    if resolution_hardcoded is not None:
        px2um = resolution_hardcoded

    xtcav_params = {
        "resolution": px2um,
        "dispersion": dispersion * 1e-6 * -1,
        "streak": hard_code_streak,
        "energy": mean_energy,
        "cavity_frequency": 11.424e9,
    }

    img_in = np.asarray(image, dtype=float)
    if bk_subtract and (bk_timestamp is not None):
        populate_correlation_entries(corr_struct_sorted, beam_time_dir, bk_timestamp)
        bk_charges, bk_images, *_ = get_charge_and_image_by_timestamp(corr_struct_sorted, bk_timestamp)
        img_in = img_in - np.mean(bk_images, axis=0)

    px2fs, px2mev, *_ = get_tcavAxes_conversion_ticks(
        xtcav_params["energy"],
        xtcav_params["streak"],
        xtcav_params["dispersion"],
        xtcav_params["cavity_frequency"],
        xtcav_params["resolution"],
        3e8,
    )

    filtered = process_tcav_image_v1(img_in[:, :], threshold=threshold)

    img = np.asarray(filtered.T, dtype=float)
    img = np.maximum(img, 0.0)

    h, w = img.shape
    img_sum = np.sum(img)
    if img_sum < image_summation_threshold:
        return {"ok": False, "error": "sum_below_threshold", "img_sum": float(img_sum)}

    # normalized image for moment calculations
    img_scaled = img / img_sum

    x_phys = (np.arange(w) - w / 2) * px2fs
    y_phys = (np.arange(h) - h / 2) * px2mev

    # current profile
    scaler_xproj = np.abs(charge_pc * 1e-12 / (np.sum(img_scaled) * px2fs * 1e-15))
    x_projection = np.sum(img_scaled * scaler_xproj, axis=0)
    xp = moving_average(x_projection, moving_average_window)
    pad = (x_projection.shape[0] - xp.shape[0]) // 2
    xp = np.pad(xp, (pad, x_projection.shape[0] - xp.shape[0] - pad), mode="edge")
    I = _nanfill_1d(xp)

    # first moment
    EPS_COL = 1e-15
    col_sum = img_scaled.sum(axis=0)
    col_sum = np.where(col_sum < EPS_COL, np.nan, col_sum)
    E_com = (img_scaled.T @ y_phys) / col_sum
    E_com = _nanfill_1d(E_com)

    # second central moment
    dy = y_phys[:, None] - E_com[None, :]
    sigmaE2 = np.sum(img_scaled * (dy ** 2), axis=0) / col_sum
    sigmaE2 = _nanfill_1d(sigmaE2)
    sigmaE2 = np.maximum(sigmaE2, 0.0)
    sigmaE = np.sqrt(sigmaE2)

    img_norm = img_scaled / max(1e-12, np.max(img_scaled))

    return {
        "ok": True,
        "img2d_use": img_scaled,
        "img_norm": img_norm,
        "x_phys": x_phys / scale,
        "y_phys": y_phys,
        "E_com": E_com,
        "sigmaE2": sigmaE2,
        "sigmaE": sigmaE,
        "I": I,
        "xp": xp,
        "px2fs": px2fs,
        "px2mev": px2mev,
    }

def recompute_moments_from_img(img2d_use, x_fs, y_mev, charge_pc, moving_average_window):
    """
    Recompute I(t), E_com(t), sigmaE2(t), sigmaE(t) after a time shift is applied to the 2D image.
    """
    xproj = np.sum(img2d_use, axis=0)
    xproj_s = moving_average(xproj, moving_average_window)
    pad = (xproj.shape[0] - xproj_s.shape[0]) // 2
    xproj_s = np.pad(xproj_s, (pad, xproj.shape[0] - xproj_s.shape[0] - pad), mode="edge")

    dt_s = float(np.median(np.diff(x_fs)) * 1e-15)
    Q_C = float(charge_pc) * 1e-12
    I_A = xproj_s * (Q_C / max(1e-30, np.sum(xproj_s) * dt_s))
    I_A = _nanfill_1d(I_A)

    col_sum = np.sum(img2d_use, axis=0)
    col_sum = np.where(col_sum < 1e-18, np.nan, col_sum)

    Ecom = (img2d_use.T @ y_mev) / col_sum
    Ecom = _nanfill_1d(Ecom)

    dy = y_mev[:, None] - Ecom[None, :]
    sigmaE2 = np.sum(img2d_use * (dy ** 2), axis=0) / col_sum
    sigmaE2 = _nanfill_1d(sigmaE2)
    sigmaE2 = np.maximum(sigmaE2, 0.0)
    sigmaE = np.sqrt(sigmaE2)

    return I_A, Ecom, sigmaE2, sigmaE


# ============================================================
# Reconstruction engines
# ============================================================
def reconstruct_delta_offset_corrected(
    x_fs,
    E_on,
    E_off,
    I_on,
    mask,
    target_J,
):
    """
    DELTA-method with additive offset correction following the SLAC slide.

    Raw:
      P_raw(t) = [E_off(t) - E_on(t)] * I_on(t) / 1000   [GW]

    Compute missing energy:
      U_offset = U_GDET - ∫ P_raw(t) dt

    Convert to an energy-per-electron shift:
      dE_shift_MeV = U_offset / (Q_window * 1e6)

    Correct the ON curve:
      E_on_corr(t) = E_on(t) - dE_shift_MeV

    Recompute:
      P_corr(t) = [E_off(t) - E_on_corr(t)] * I_on(t) / 1000
    """
    x_fs = np.asarray(x_fs, dtype=float)
    E_on = np.asarray(E_on, dtype=float)
    E_off = np.asarray(E_off, dtype=float)
    I_on = np.asarray(I_on, dtype=float)
    mask = np.asarray(mask, dtype=bool)

    # raw delta method
    dE_raw = E_off - E_on                       # MeV
    P_raw_GW = dE_raw * I_on / 1000.0          # GW

    # integrated raw reconstructed X-ray energy
    U_raw_J = integrate_power_J(x_fs, P_raw_GW, mask=mask)

    # integrated beam charge over reconstruction window
    Q_window_C = integrate_current_C(x_fs, I_on, mask=mask)

    # missing energy in Joules
    if not np.isfinite(U_raw_J) or not np.isfinite(Q_window_C) or Q_window_C <= 0 or target_J is None:
        U_offset_J = np.nan
        dE_shift_MeV = np.nan
        E_on_corr = E_on.copy()
    else:
        U_offset_J = target_J - U_raw_J

        # convert Joules/electron -> eV, then -> MeV
        # since 1 eV = 1 J/C for one electron charge basis, divide by charge in Coulombs
        dE_shift_MeV = U_offset_J / Q_window_C / 1e6

        # shift ON curve exactly as in slide
        E_on_corr = E_on - dE_shift_MeV

    # recompute corrected delta method
    dE_corr = E_off - E_on_corr
    P_corr_GW = dE_corr * I_on / 1000.0
    U_corr_J = integrate_power_J(x_fs, P_corr_GW, mask=mask)

    return {
        "E_on_corr_MeV": E_on_corr,
        "dE_raw_MeV": dE_raw,
        "P_raw_GW": P_raw_GW,
        "U_raw_J": U_raw_J,
        "Q_window_C": Q_window_C,
        "U_offset_J": U_offset_J,
        "dE_shift_MeV": dE_shift_MeV,
        "dE_corr_MeV": dE_corr,
        "P_corr_GW": P_corr_GW,
        "U_corr_J": U_corr_J,
    }

def reconstruct_sigma_normalized(
    x_fs,
    sigmaE2_on,
    sigmaE2_off,
    I_on,
    mask,
    target_J,
):
    """
    SIGMA-method:
      P_sigma(t) ∝ [sigma_on^2(t) - sigma_off^2(t)] * I_on(t)^(2/3)
    then multiplicatively normalize to target_J over mask.
    """
    x_fs = np.asarray(x_fs, dtype=float)
    sigmaE2_on = np.asarray(sigmaE2_on, dtype=float)
    sigmaE2_off = np.asarray(sigmaE2_off, dtype=float)
    I_on = np.asarray(I_on, dtype=float)
    mask = np.asarray(mask, dtype=bool)

    dsig2 = sigmaE2_on - sigmaE2_off
    Iterm = np.power(np.maximum(I_on, 0.0), 2.0 / 3.0)
    P_shape = dsig2 * Iterm

    P_norm, U_raw_J, scale = normalize_profile_to_energy(
        t_fs=x_fs,
        P_arb=P_shape,
        mask=mask,
        target_J=target_J,
    )

    return {
        "dsig2_MeV2": dsig2,
        "P_shape_arb": P_shape,
        "P_norm": P_norm,
        "U_raw_J": U_raw_J,
        "scale": scale,
    }
# ============================================================
# Plot helpers
# ============================================================
def xtcav_plot_from_processed(proc, title=""):
    x = np.asarray(proc["x_phys"], dtype=float)
    y = np.asarray(proc["y_phys"], dtype=float)
    img = np.asarray(proc["img_norm"], dtype=float)

    y_plot = y.copy()
    if PLOT_RECENTER_Y_FOR_DISPLAY:
        if RE_CENTER_Y_BY == "median":
            row_w = np.sum(img, axis=1)
            row_w = row_w / max(1e-12, np.sum(row_w))
            y0 = np.sum(y * row_w)
        elif RE_CENTER_Y_BY == "com":
            w = np.sum(img)
            y0 = np.sum(img * y[:, None]) / max(1e-12, w)
        else:
            raise ValueError("RE_CENTER_Y_BY must be 'median' or 'com'")
        y_plot = y_plot - y0

    fig, ax = plt.subplots(figsize=(3.0, 3.0), constrained_layout=True)
    ax.set_title(title)
    im = ax.imshow(
        img, cmap=CMAP, origin="upper",
        extent=[x[0], x[-1], y_plot[-1], y_plot[0]],
        aspect="auto", vmin=0
    )
    ax.set_xlim(*XLIM_FS)
    ax.set_ylim(*IMAGE_YLIM_MEV)
    ax.set_xlabel("Time (fs)")
    ax.set_ylabel("ΔEnergy (MeV)")
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.02)
    plt.show()
    return fig

def plot_alignment_fom(sb_grid, rhoI_curve, rhoE_curve, score_curve, sb_best, idx_best, x_ref):
    fig, ax = plt.subplots(figsize=(7.6, 4.2), constrained_layout=True)
    ax.set_title(f"Alignment FoM vs integer time-shift for best conservative OFF shot idx={idx_best}")
    ax.set_xlabel(f"Shift (bins)   [dt={np.median(np.diff(x_ref)):.2f} fs/bin ; + = OFF delayed]")
    ax.set_ylabel("Correlation (0–1)")
    ax.plot(sb_grid, rhoI_curve, lw=1.8, label="rhoI (current)")
    ax.plot(sb_grid, rhoE_curve, lw=1.4, ls="--", label="rhoE (COM energy)")
    ax.plot(sb_grid, score_curve, lw=1.4, ls=":", label="combined score")
    ax.axvline(0, lw=0.8, alpha=0.5)
    ax.axvline(sb_best, lw=1.0, alpha=0.5, label=f"best current shift = {sb_best:+d}")
    ax.legend(fontsize=9)
    plt.show()
    return fig

def plot_deltaE_panels(
    img2d_off0,
    y_mev_off,
    x_ref,
    E_on,
    charge_off_pc,
    moving_avg_window,
    mwin,
    sb_best,
    offset_bins_list,
    idx_best,
    rhoI_best,
    rhoE_best,
):
    fig, axes = plt.subplots(2, 3, figsize=(12.5, 6.5), constrained_layout=True, sharex=True, sharey=True)
    axes = axes.ravel()

    for ax, offb in zip(axes, offset_bins_list):
        sb = int(sb_best + offb)
        img2d_sh = shift_2d_time_zerofill(img2d_off0, sb)

        _, E_off_s, _, _ = recompute_moments_from_img(
            img2d_use=img2d_sh,
            x_fs=x_ref,
            y_mev=y_mev_off,
            charge_pc=charge_off_pc,
            moving_average_window=moving_avg_window,
        )

        dE = E_off_s - E_on
        ax.plot(x_ref, dE, lw=1.8)
        ax.axhline(0, ls="--", lw=0.8, alpha=0.7)
        if mwin.sum() > 1:
            ax.axvspan(x_ref[mwin][0], x_ref[mwin][-1], alpha=0.08)
        ax.set_title(f"shift = {sb:+d} bins  (best {sb_best:+d} + {offb:+d})")
        ax.set_xlim(*XLIM_FS)

    for ax in axes[3:]:
        ax.set_xlabel("Time (fs)")
    for ax in axes[::3]:
        ax.set_ylabel("Δ⟨E⟩ (MeV)")

    fig.suptitle(
        f"Δ⟨E⟩(t) = ⟨E_off⟩ - ⟨E_on⟩ for best OFF idx={idx_best} (current-aligned)\n"
        f"Best shift = {sb_best:+d} bins ({sb_best*np.median(np.diff(x_ref)):+.1f} fs), "
        f"rhoI={rhoI_best:.3f}, rhoE={rhoE_best:.3f}",
        y=1.02
    )
    plt.show()
    return fig
```

```python

def initialize_diag0_525_struct(folder):
    entries = []
    pattern = re.compile(
        r"(?P<type>[\w\-]+)-(?P<date>\d{4}-\d{2}-\d{2})-(?P<hh>\d{2})(?P<mm>\d{2})(?P<ss>\d{2})\.mat",
        re.IGNORECASE
    )

    for fname in os.listdir(folder):
        # Match only `.mat` files that contain "diag0_525" (case-insensitive)
        if ".mat" not in fname.lower() or "diag0_525" not in fname.lower():
            continue

        match = pattern.search(fname)
        if match:
            hhmmss = int(match["hh"] + match["mm"] + match["ss"])
            entries.append((
                match["type"],     # type field
                None,              # instr (not present in this pattern)
                match["date"],     # date
                hhmmss,            # time as integer
                fname,             # filename
                None,              # charge array placeholder
                False,             # valid_charge flag
                None,              # image array placeholder
                False               # valid_image flag
            ))

    return np.array(entries, dtype=correlation_dtype)

def get_diag0_525_res(folder, target_time):
    """
    Find the DIAG0_525 .mat file with a specific HHMMSS timestamp 
    and return the 'res' value from the 'data' group.

    Args:
        folder (str): Path to the directory containing .mat files
        target_time (str or int): Time in HHMMSS format, e.g., '105656'

    Returns:
        float or np.ndarray: The resolution value from the file
    """
    target_time = str(target_time)  # ensure string comparison
    
    pattern = re.compile(
        r"(?P<type>[\w\-]+)-(?P<instr>[\w_]+)-(?P<date>\d{4}-\d{2}-\d{2})-(?P<hh>\d{2})(?P<mm>\d{2})(?P<ss>\d{2})\.mat",
        re.IGNORECASE
    )

    for fname in os.listdir(folder):
        if "diag0_525" not in fname.lower() or not fname.lower().endswith(".mat"):
            continue
        
        match = pattern.match(fname)
        if match:
            hhmmss = match["hh"] + match["mm"] + match["ss"]
            if hhmmss == target_time:
                file_path = os.path.join(folder, fname)
                with h5py.File(file_path, "r") as f:
                    return f["data"]["res"][0]

    raise FileNotFoundError(f"No DIAG0_525 file with time {target_time} found in {folder}")
```

```python
def prep_stcav_v2(
    target_timestamp,
    stcav_params,
    *,
    corr_struct_sorted,
    beam_time_dir,

    image_summation_threshold=1000,
    moving_average_window=30,
    y_moving_average_window=None,
    threshold=5,
    every_other_plot_number=1,

    bk_subtract=False,
    bk_timestamp=None,

    # --- auto-exclusion on UNCENTERED projection (same knobs as xtcav) ---
    recommend_excludes=True,
    detect_debug=False,
    detect_debug_max_print=10,
    detect_edge_bins_frac=0.10,
    detect_edge_bins_min=10,
    detect_clip_fraction_threshold=0.18,
    detect_proj_baseline_percentile=5.0,
    detect_require_asymmetry=False,
    detect_asymmetry_ratio_threshold=1.5,
    detect_robust_stat="mean",

    # --- plotting (same idea as xtcav) ---
    plot_uncentered=True,
    color="jet",
    x_lim=(-400, 400),   # in fs (assuming px2fs from get_tcavAxes_conversion_ticks)
    y_lim=(10, 50),      # in MeV
):
    """
    STCAV twin of prep_xtcav:
      returns (proj_list, charge_list, idx_list, x_phys,
               recommended_exclude_raw, recommended_meta, index_map)

    Assumptions:
      - stcav_params has keys: energy, streak, dispersion, cavity_frequency, resolution
      - get_tcavAxes_conversion_ticks returns px2fs, px2mev consistent with xtcav usage
    """

    if y_moving_average_window is None:
        y_moving_average_window = moving_average_window

    # # --- load shots (identical pattern to xtcav) ---
    # charges, images, valid_charge, valid_image = load_tcav_shots(
    #     target_timestamp=target_timestamp,
    #     corr_struct_sorted=corr_struct_sorted,
    #     beam_time_dir=beam_time_dir,
    #     bk_subtract=bk_subtract,
    #     bk_timestamp=bk_timestamp,
    #     dtype=np.float32,
    # )

    # --- load shots (same as prep_xtcav) ---
    populate_correlation_entries(corr_struct_sorted, beam_time_dir, target_timestamp)
    charges, images, valid_charge, valid_image = get_charge_and_image_by_timestamp(
        corr_struct_sorted, target_timestamp
    )

    # --- optional background subtraction (same as prep_xtcav) ---
    if bk_subtract and (bk_timestamp is not None):
        populate_correlation_entries(corr_struct_sorted, beam_time_dir, bk_timestamp)
        _, bk_images, _, _ = get_charge_and_image_by_timestamp(corr_struct_sorted, bk_timestamp)
        images = images - np.mean(bk_images, axis=0)

    # --- calibration (same callsite) ---
    px2fs, px2mev, xticks, yticks = get_tcavAxes_conversion_ticks(
        stcav_params["energy"],
        stcav_params["streak"],
        stcav_params["dispersion"],
        stcav_params["cavity_frequency"],
        stcav_params["resolution"],
        3e8,
    )
    print("px2fs")
    print(px2fs)
    print("px2mev")
    print(px2mev)

    # --- sorting by charge defines projection index space ---
    sorted_indices = np.argsort(charges)

    proj_list = []
    charge_list = []
    idx_list = []

    recommended_exclude_raw = []
    recommended_meta = {}

    # debug counters (same behavior)
    n_print_flag = 0
    n_print_ok = 0

    m = 0
    x_phys = None

    for i in sorted_indices:
        # 1) morphology/threshold pipeline (same)
        filtered = process_tcav_image_v1(images[i, :, :], threshold=threshold)

        # uncentered representation used for detection + optional plotting
        img_uncentered = filtered.T  # shape (h, w)

        # 2) sum gate (same)
        img_sum = float(np.sum(img_uncentered))
        if img_sum < image_summation_threshold:
            print(f"Image does not contain a signal; Index {i}")
            continue

        # 3) detect clipping on uncentered projection (same)
        clipped = False
        info = None
        if recommend_excludes:
            debug_this = False
            if detect_debug:
                debug_this = (n_print_flag < detect_debug_max_print) or (n_print_ok < detect_debug_max_print)

            clipped, info = detect_edge_clipping_from_uncentered_projection(
                img2d_uncentered=img_uncentered,
                edge_bins_frac=detect_edge_bins_frac,
                edge_bins_min=detect_edge_bins_min,
                clip_fraction_threshold=detect_clip_fraction_threshold,
                proj_baseline_percentile=detect_proj_baseline_percentile,
                require_asymmetry=detect_require_asymmetry,
                asymmetry_ratio_threshold=detect_asymmetry_ratio_threshold,
                robust_stat=detect_robust_stat,
                debug=debug_this,
                debug_id=f"raw_idx={i}",
            )

            if debug_this:
                if clipped:
                    n_print_flag += 1
                else:
                    n_print_ok += 1

            if clipped:
                recommended_exclude_raw.append(int(i))
                recommended_meta[int(i)] = info

        # 4) scale by charge (same)
        charge_i = float(charges[i])
        img_normalized_uncentered = (img_uncentered / img_sum)
        scaler_xproj = charge_i * 1e-12/(np.sum(img_normalized_uncentered) * px2fs * 1e-15)
        scaler_yproj = charge_i * 1e-12/(np.sum(img_normalized_uncentered) * px2mev * 1e6)

        h, w = img_normalized_uncentered.shape

        # centroid for recentering (same)
        x_idx = np.arange(w)
        y_idx = np.arange(h)
        total = img_normalized_uncentered.sum()
        x_c = (img_normalized_uncentered.sum(axis=0) @ x_idx) / total
        y_c = (img_normalized_uncentered.sum(axis=1) @ y_idx) / total
        dx = int(round(w / 2 - x_c))
        dy = int(round(h / 2 - y_c))

        # recenter for projections (same)
        img_centered = np.roll(np.roll(img_normalized_uncentered, shift=dy, axis=0), shift=dx, axis=1)

        # physical axes (centered convention) (same)
        x_phys = (np.arange(w) - w / 2) * px2fs
        y_phys = (np.arange(h) - h / 2) * px2mev

        # x projection from centered image (same)
        x_projection = np.sum(img_centered*scaler_xproj, axis=0)
        x_projection_smooth = moving_average(x_projection, moving_average_window)
        pad = (x_projection.shape[0] - x_projection_smooth.shape[0]) // 2
        x_projection_smooth_padded = np.pad(
            x_projection_smooth,
            (pad, x_projection.shape[0] - x_projection_smooth.shape[0] - pad),
            mode="edge",
        )

        # y projection (same)
        y_projection = np.sum(img_centered*scaler_yproj, axis=1)
        y_projection_smooth = moving_average(y_projection, y_moving_average_window)
        pad = (y_projection.shape[0] - y_projection_smooth.shape[0]) // 2
        y_projection_smooth_padded = np.pad(
            y_projection_smooth,
            (pad, y_projection.shape[0] - y_projection_smooth.shape[0] - pad),
            mode="edge",
        )

        # optional plotting (same concept as xtcav)
        if plot_uncentered and (m == 0):
            proj_idx = len(proj_list)

            fig, ax = plt.subplots(figsize=(8, 6))

            x_phys_unc = (np.arange(w) - w / 2) * px2fs/1000 #put in ps
            y_phys_unc = (np.arange(h) - h / 2) * px2mev

            im = ax.imshow(
                img_normalized_uncentered,
                cmap=color,
                origin="upper",
                extent=[x_phys_unc[0], x_phys_unc[-1], y_phys_unc[-1], y_phys_unc[0]],
                aspect="auto",
                rasterized=False,
            )

            ax.set_title(f"raw_idx {i} | proj_idx {proj_idx} | charge={charge_i:.3f} | clipped={clipped}")
            ax.set_xlim(*x_lim)
            ax.set_ylim(*y_lim)
            ax.set_xlabel("Time (ps)")
            ax.set_ylabel("Delta Energy (MeV)")

            # overlay projection used for detection
            proj_bs = info["proj_bs"] if (info is not None and "proj_bs" in info) else img_uncentered.sum(axis=0)
            proj_bs_n = proj_bs / (np.max(proj_bs) + 1e-12)

            axr = ax.twinx()
            axr.plot(x_phys_unc, proj_bs_n, color="red", lw=0.8, alpha=0.9)
            axr.set_ylabel("Detection proj (norm)", color="red")
            axr.tick_params(axis="y", colors="red")
            axr.set_ylim(0, 1.05)

            cbar = fig.colorbar(im, ax=ax, pad=0.02, fraction=0.05)
            cbar.set_label("Intensity (a.u.)")
            plt.show()

        m += 1
        if m > every_other_plot_number:
            m = 0

        proj_list.append(x_projection_smooth_padded.astype(float))
        charge_list.append(charge_i)
        idx_list.append(int(i))

    recommended_exclude_raw = sorted(set(recommended_exclude_raw))

    proj_to_raw = {k: int(raw) for k, raw in enumerate(idx_list)}
    raw_to_proj = {int(raw): k for k, raw in enumerate(idx_list)}
    index_map = {"proj_to_raw": proj_to_raw, "raw_to_proj": raw_to_proj}

    return (
        proj_list,
        charge_list,
        idx_list,
        x_phys,
        recommended_exclude_raw,
        recommended_meta,
        index_map,
    )
```

```python
def plot_stcav_four(
    target_timestamp,
    stcav_params,
    indices,
    save_name="stcav_four",
    image_summation_threshold=1000,
    moving_average_window=30,
    y_moving_average_window=None,
    threshold=5,
    bk_subtract=False,
    bk_timestamp=None,
    color="jet",
    y_lim=(-30, 30),
    x_lim=(-200, 200),
    axis_fs_ps="fs",
):
    """
    STCAV version of plot_xtcav_four:
      - no TREX
      - stcav_params provided directly
      - uses same load/bk-subtract pattern as prep_xtcav
      - indices are RAW indices into images[i,:,:]
    """
    assert len(indices) == 4, "Please pass exactly four indices."

    if y_moving_average_window is None:
        y_moving_average_window = moving_average_window

    if axis_fs_ps == "ps":
        scale = 1000
        x_axis_label = "Time (ps)"
    else:
        scale = 1
        x_axis_label = "Time (fs)"

    # --- load shots (same pattern as prep_xtcav) ---
    populate_correlation_entries(corr_struct_sorted, beam_time_dir, target_timestamp)
    charges, images, valid_charge, valid_image = get_charge_and_image_by_timestamp(
        corr_struct_sorted, target_timestamp
    )

    # --- optional background subtraction (same pattern as prep_xtcav) ---
    if bk_subtract and (bk_timestamp is not None):
        populate_correlation_entries(corr_struct_sorted, beam_time_dir, bk_timestamp)
        _, bk_images, _, _ = get_charge_and_image_by_timestamp(corr_struct_sorted, bk_timestamp)
        images = images - np.mean(bk_images, axis=0)

    # --- calibration (same call as xtcav) ---
    px2fs, px2mev, xticks, yticks = get_tcavAxes_conversion_ticks(
        stcav_params["energy"],
        stcav_params["streak"],
        stcav_params["dispersion"],
        stcav_params["cavity_frequency"],
        stcav_params["resolution"],
        3e8,
    )
    print("Axis info")
    print("px2fs");  print(px2fs)
    print("px2mev"); print(px2mev)

    def process_one(i):
        if i < 0 or i >= images.shape[0]:
            return None, None, None, None, None, f"Index {i} out of range (0..{images.shape[0]-1})."

        filtered = process_tcav_image_v1(images[i, :, :], threshold=threshold)
        img = filtered.T
        h, w = img.shape
        img_sum = float(np.sum(img))
        if img_sum < image_summation_threshold:
            return None, None, None, None, None, f"Skipping index {i}: sum<{image_summation_threshold}"

        charge = float(charges[i])
        img_scaled = (img / img_sum)

        scaler_xproj = np.abs(charge * 1e-12/(np.sum(img_scaled) * px2fs * 1e-15))
        scaler_yproj = np.abs(charge * 1e-12/(np.sum(img_scaled) * px2mev * 1e6))

        # centroid recenter (integer)
        x_idx = np.arange(w)
        y_idx = np.arange(h)
        total = img_scaled.sum()
        x_c = (img_scaled.sum(axis=0) @ x_idx) / total
        y_c = (img_scaled.sum(axis=1) @ y_idx) / total
        dx = int(round(w / 2 - x_c))
        dy = int(round(h / 2 - y_c))
        img_centered = np.roll(np.roll(img_scaled, shift=dy, axis=0), shift=dx, axis=1)

        # physical axes (centered)
        x_phys = (np.arange(w) - w / 2) * px2fs
        y_phys = (np.arange(h) - h / 2) * px2mev

        # projections + smoothing + padding
        x_projection = np.sum(img_centered*scaler_xproj, axis=0)
        xp = moving_average(x_projection, moving_average_window)
        pad = (x_projection.shape[0] - xp.shape[0]) // 2
        xp = np.pad(xp, (pad, x_projection.shape[0] - xp.shape[0] - pad), mode="edge")

        y_projection = np.sum(img_centered*scaler_yproj, axis=1)
        yp = moving_average(y_projection, y_moving_average_window)
        pad = (y_projection.shape[0] - yp.shape[0]) // 2
        yp = np.pad(yp, (pad, y_projection.shape[0] - yp.shape[0] - pad), mode="edge")

        return img_centered, x_phys, y_phys, (xp, yp), charge, None

    processed = []
    for i in indices:
        processed.append((i, *process_one(i)))

    fig, axes = plt.subplots(2, 2, figsize=(12, 9), constrained_layout=True)
    axes = axes.ravel()

    for ax, (i, img_centered, x_phys, y_phys, projs, charge_i, err) in zip(axes, processed):
        ax.set_title(f"idx {i} | charge={charge_i:.3f}" if charge_i is not None else f"idx {i}")

        if err is not None or img_centered is None:
            ax.text(0.5, 0.5, err if err else "No data", ha="center", va="center", transform=ax.transAxes)
            ax.axis("off")
            continue

        im = ax.imshow(
            img_centered/np.max(img_centered),
            cmap=color,
            origin="upper",
            extent=[x_phys[0] / scale, x_phys[-1] / scale, y_phys[-1], y_phys[0]],
            aspect="auto",
            vmin=0,
        )

        ax.axhline(0, ls="--", color="black", lw=0.8)
        ax.axvline(0, ls="--", color="black", lw=0.8)
        ax.set_xlim(x_lim[0], x_lim[1])
        ax.set_ylim(y_lim[0], y_lim[1])
        ax.set_xlabel(x_axis_label)
        ax.set_ylabel("ΔEnergy (MeV)")

        # per-panel colorbar
        cbar = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.02)
        cbar.set_label("Intensity (a.u.)", fontsize=9)

        # overlay x-projection on right axis
        xp, yp = projs
        axr = ax.twinx()
        axr.plot(x_phys / scale, xp, color="red", lw=0.6, alpha=0.95)
        axr.set_ylabel("Current (A)", color="red")
        axr.tick_params(axis="y", colors="red", labelsize=8)
        axr.spines["right"].set_edgecolor("red")
        axr.set_xlim(ax.get_xlim())
        axr.set_ylim(0, 1.1 * max(1e-12, xp.max()))

        # top axis for y projection
        axt = ax.twiny()
        axt.plot(yp/np.max(yp), y_phys, color="red", lw=0.6, alpha=0.95)
        axt.set_xlabel("Y proj (a.u.)", color="red")
        axt.tick_params(axis="x", colors="red", labelsize=8)
        axt.spines["top"].set_edgecolor("red")
        # axt.set_ylim(ax.get_ylim())
        # axt.set_xlim(0, 1.1 * max(1e-12, yp.max()))

    if save_name:
        fig.savefig(f"{save_name}.pdf", dpi=300)
    plt.show()

    return {"indices": indices, "processed": processed, "px_scale": (px2fs, px2mev)}

def plot_stcav_four_proj(
    target_timestamp,
    stcav_params,
    proj_indices,
    index_map,
    **kwargs,
):
    """
    proj_indices: indices into proj_list (order produced by prep_stcav_v2)
    index_map: output from prep_stcav_v2 (contains 'proj_to_raw')
    kwargs: forwarded to plot_stcav_four (thresholds, bk_subtract, etc.)
    """
    assert len(proj_indices) == 4, "Pass exactly four projection indices."
    proj_to_raw = index_map["proj_to_raw"]

    raw_indices = []
    for k in proj_indices:
        if k not in proj_to_raw:
            raise IndexError(f"Projection index {k} not found. Valid range is 0..{len(proj_to_raw)-1}.")
        raw_indices.append(int(proj_to_raw[k]))

    return plot_stcav_four(
        target_timestamp=target_timestamp,
        stcav_params=stcav_params,
        indices=raw_indices,
        **kwargs,
    )
```

```python
# ============================================================
# XRAY Lasing
# ============================================================
#TREX
time_stamp = 144459
trex_time_stamp = 144500
streak = 7.37
moving_avg_window = 17
y_moving_avg = 10
bk_timestamp = 145451
image_summation_threshold = 2000
resolution_hardcoded = 29.2

bg = estimate_xtcav_background_stats(
    target_timestamp=time_stamp,
    bk_subtract=True,
    bk_timestamp=145451,
    roi_x=(924,1024),
    roi_y=(0,100),
    robust=True
)

n_sigma = 5.0
threshold = bg["median"] + n_sigma * bg["sigma_mad"]
print("threshold:", threshold)

charges, images, valid_charge, valid_image = get_charge_and_image_by_timestamp(corr_struct_sorted, time_stamp)
target_timestamp = 141909
filtered_entries = trex_struct_sorted[trex_struct_sorted['timestamp'] == target_timestamp]
print(filtered_entries)
filename = filtered_entries['filename'][0]
print(filename)
trex = beam_time_dir + filename
print(trex)
trex_mediumDeep_lasing_file = h5py.File(trex)

lasing_shot = trex_mediumDeep_lasing_file['data']['img_orig'][()]
lasing_shot_charge = float(trex_mediumDeep_lasing_file['data']['charge'][()][0][0])*1e12
# ----------------------------
# USER INPUTS
# ----------------------------
on_image = lasing_shot
on_charge = lasing_shot_charge
on_trex_stamp = trex_time_stamp

off_images = images
off_charges = charges
off_trex_stamps = trex_time_stamp   # scalar OK (same stamp for all OFF)

streak = streak
moving_avg_window = moving_avg_window
y_moving_avg = y_moving_avg
threshold = threshold
bk_subtract = True
bk_timestamp = bk_timestamp
resolution_hardcoded = resolution_hardcoded

# Conservative OFF-shot selection
TOPK          = 5
CHARGE_TOL_PC = 3.0
CHARGE_UNITS  = "pC"   # "pC" or "C"

# Correlation weighting for conservative search
USE_CURRENT_IN_SCORE = True
CURRENT_SCORE_WEIGHT = 0.35
ENERGY_SCORE_WEIGHT  = 1.0 - CURRENT_SCORE_WEIGHT  # 0.65

# ON-defined trusted window
CURRENT_FRACTION_GATE  = 0.05
MIN_WINDOW_SAMPLES     = 20
FALLBACK_TIME_WINDOWFS = (-100.0, 100.0)

# Local shift study
MAX_LOCAL_SHIFT_BINS      = 12
SHIFT_PANEL_OFFSETS       = [-5, -3, -1, 0, +1, +3]
POWER_ENVELOPE_SHIFT_SPAN = 5

# Plot controls
XLIM_FS        = (-200.0, 200.0)
IMAGE_YLIM_MEV = (-20.0, 20.0)
CMAP = "jet"

# Display-only Y recenter for images
PLOT_RECENTER_Y_FOR_DISPLAY = True
RE_CENTER_Y_BY = "median"  # "median" or "com"


SAVE_DIR = save_name_initial+"/plots_xtcav_dualmethod"
SAVE_TAG = "xtcav_dualmethod"


if not isinstance(trex, (str, os.PathLike)):
    raise TypeError(
        f"`trex` must be the TREX calibration H5 filepath (str/PathLike). Got {type(trex)}. "
        "Do NOT set trex = trex_time_stamp."
    )
trex_path = os.fspath(trex)

on_proc = process_single_xtcav_with_moments(
    on_image, charge_to_pC(on_charge), trex_path,
    hard_code_streak=streak,
    image_summation_threshold=1,
    moving_average_window=moving_avg_window,
    y_moving_average_window=y_moving_avg,
    threshold=threshold,
    bk_subtract=bk_subtract,
    bk_timestamp=bk_timestamp,
    resolution_hardcoded=resolution_hardcoded,
)
if not on_proc.get("ok", False):
    raise RuntimeError(f"ON failed: {on_proc}")

x_ref      = np.asarray(on_proc["x_phys"], dtype=float)
E_on       = np.asarray(on_proc["E_com"], dtype=float)
sigmaE2_on = np.asarray(on_proc["sigmaE2"], dtype=float)
I_on       = np.asarray(on_proc["I"], dtype=float)

thr = CURRENT_FRACTION_GATE * np.nanmax(I_on)
mwin = (I_on >= thr)
if mwin.sum() < MIN_WINDOW_SAMPLES:
    mwin = (x_ref >= FALLBACK_TIME_WINDOWFS[0]) & (x_ref <= FALLBACK_TIME_WINDOWFS[1])

fig_on = xtcav_plot_from_processed(on_proc, title="ON (lasing)")
maybe_savefig(fig_on, f"{SAVE_TAG}__ON_lasing")

q_on_pc  = float(charge_to_pC(on_charge))
q_off_pc = charge_to_pC(off_charges)
eligible = np.where(np.abs(q_off_pc - q_on_pc) <= CHARGE_TOL_PC)[0]

print(f"Charge gate: Q_on={q_on_pc:.3f} pC, tol=±{CHARGE_TOL_PC:.3f} pC -> eligible OFF: {len(eligible)}/{len(off_images)}")
print(f"Conservative score weights: current={CURRENT_SCORE_WEIGHT:.2f}, energy={ENERGY_SCORE_WEIGHT:.2f}")

scores        = np.full(len(off_images), -np.inf, dtype=float)
proc_cache    = [None] * len(off_images)
E_off_all     = [None] * len(off_images)
I_off_all     = [None] * len(off_images)
sigmaE2_all   = [None] * len(off_images)

for i in eligible:
    proc = process_single_xtcav_with_moments(
        off_images[i], off_charges[i], trex_path,
        hard_code_streak=streak,
        image_summation_threshold=1,
        moving_average_window=moving_avg_window,
        y_moving_average_window=y_moving_avg,
        threshold=threshold,
        bk_subtract=bk_subtract,
        bk_timestamp=bk_timestamp,
        resolution_hardcoded=resolution_hardcoded,
    )
    if not proc.get("ok", False):
        continue

    E_off = _interp_to(proc["x_phys"], proc["E_com"], x_ref)
    I_off = _interp_to(proc["x_phys"], proc["I"], x_ref)
    sig2  = _interp_to(proc["x_phys"], proc["sigmaE2"], x_ref)

    sE = corr01(E_off[mwin], E_on[mwin])
    if USE_CURRENT_IN_SCORE:
        sI = corr01(I_off[mwin], I_on[mwin])
        sT = ENERGY_SCORE_WEIGHT * sE + CURRENT_SCORE_WEIGHT * sI
    else:
        sT = sE

    scores[i]      = sT
    proc_cache[i]  = proc
    E_off_all[i]   = E_off
    I_off_all[i]   = I_off
    sigmaE2_all[i] = sig2

valid = np.isfinite(scores) & (scores > -np.inf)
if valid.sum() == 0:
    raise RuntimeError("No valid OFF shots after conservative search.")

top_idx = np.argsort(scores[valid])[::-1][:TOPK]
top_idx = np.where(valid)[0][top_idx]

print("\nTop candidates from conservative search:")
for r, idx in enumerate(top_idx, 1):
    print(f"  {r:>2d}) idx={idx:>5d}  score={scores[idx]:.4f}  Q={q_off_pc[idx]:.3f} pC")

# show matched OFF shots
for idx in top_idx:
    fig_off = xtcav_plot_from_processed(proc_cache[idx], title=f"OFF idx={idx} score={scores[idx]:.3f}")
    maybe_savefig(fig_off, f"{SAVE_TAG}__OFF_idx{idx:04d}_score{scores[idx]:.3f}")

# overlays for conservative search
figE, axE = plt.subplots(figsize=(7.2, 3.6), constrained_layout=True)
axE.set_title("E_COM(t) overlay (ON vs top OFF) — conservative search")
axE.set_xlabel("Time (fs)")
axE.set_ylabel("COM energy (MeV)")
axE.plot(x_ref, np.where(mwin, E_on, np.nan), lw=2.0, label="ON (lasing)")
for idx in top_idx:
    axE.plot(x_ref, np.where(mwin, E_off_all[idx], np.nan), lw=1.2, alpha=0.9, label=f"OFF idx={idx}")
axE.set_xlim(*XLIM_FS)
axE.legend(fontsize=8, ncol=2)
plt.show()
maybe_savefig(figE, f"{SAVE_TAG}__Ecom_overlay_conservative_top{TOPK}")

figI, axI = plt.subplots(figsize=(7.2, 3.6), constrained_layout=True)
axI.set_title("I(t) overlay (ON vs top OFF) — conservative search")
axI.set_xlabel("Time (fs)")
axI.set_ylabel("Current (A)")
axI.plot(x_ref, np.where(mwin, I_on, np.nan), lw=2.0, label="ON (lasing)")
for idx in top_idx:
    axI.plot(x_ref, np.where(mwin, I_off_all[idx], np.nan), lw=1.2, alpha=0.9, label=f"OFF idx={idx}")
axI.set_xlim(*XLIM_FS)
axI.legend(fontsize=8, ncol=2)
plt.show()
maybe_savefig(figI, f"{SAVE_TAG}__I_overlay_conservative_top{TOPK}")
```
